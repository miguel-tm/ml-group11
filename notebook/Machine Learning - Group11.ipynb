
## 3. Pre-Split Data Cleaning

Perform only operations that do **not** learn from the dataset and do not use the target variable.

#### 3.1 Schema & Types
- Rename columns, set data types, standardize units (e.g., BMI), parse dates.

#### 3.2 Deterministic Cleaning
- Trim strings, fix coding errors (e.g., “Yes/No” → 1/0), normalize category labels, drop duplicate rows.

#### 3.3 Domain-Rule Outlier Caps (Optional)
- Apply clinical/domain limits (e.g., BMI ∈ [10, 80]) using fixed rules.

### 3.4 Target Column Checkup
- Check if the Target Column format and contents are as expected

> **Purpose**: Improve data quality without introducing leakage and ensure reproducibility.


```python
# ============================================================================
# STEP 3: PRE-SPLIT DATA CLEANING
# ============================================================================
print("=" * 80)
print("STEP 3: PRE-SPLIT DATA CLEANING")
print("=" * 80)
print("\nPerforming operations that do NOT learn from the dataset and do NOT use the target variable.")
print("This ensures no data leakage and maintains reproducibility.\n")

# Create a copy to preserve original data
df_clean = df.copy()

# ============================================================================
# 3.1 Schema & Types
# ============================================================================
print("=" * 80)
print("3.1 SCHEMA & TYPES")
print("=" * 80)

# Check current data types
print("\nCurrent Data Types:")
print("-" * 80)
print(df_clean.dtypes)

# Based on the data dictionary, most features are binary (0/1) or integers
# Let's verify and convert to appropriate types

# Define expected data types based on documentation
binary_features = [
    'Diabetes_binary', 'HighBP', 'HighChol', 'CholCheck', 'Smoker', 
    'Stroke', 'HeartDiseaseorAttack', 'PhysActivity', 'Fruits', 
    'Veggies', 'HvyAlcoholConsump', 'AnyHealthcare', 'NoDocbcCost', 
    'DiffWalk', 'Sex'
]

integer_features = ['BMI', 'GenHlth', 'MentHlth', 'PhysHlth', 'Age', 'Education', 'Income']

# Convert to integer type (they're currently float64)
print("\n✓ Converting features to appropriate integer types...")
for col in binary_features + integer_features:
    if col in df_clean.columns:
        df_clean[col] = df_clean[col].astype('int64')

print("\nUpdated Data Types:")
print("-" * 80)
print(df_clean.dtypes)

# ============================================================================
# 3.2 Deterministic Cleaning
# ============================================================================
print("\n" + "=" * 80)
print("3.2 DETERMINISTIC CLEANING")
print("=" * 80)

# Check for duplicate rows
print("\nChecking for duplicate rows...")
duplicates = df_clean.duplicated().sum()
print(f"Number of duplicate rows: {duplicates:,}")

if duplicates > 0:
    print(f"✓ Removing {duplicates:,} duplicate rows...")
    df_clean = df_clean.drop_duplicates()
    print(f"New dataset shape: {df_clean.shape[0]:,} rows × {df_clean.shape[1]} columns")
else:
    print("✓ No duplicate rows found.")

# Check for missing values
print("\nChecking for missing values...")
missing_values = df_clean.isnull().sum()
total_missing = missing_values.sum()
print(f"Total missing values: {total_missing:,}")

if total_missing > 0:
    print("\nMissing values by column:")
    print("-" * 80)
    print(missing_values[missing_values > 0])
else:
    print("✓ No missing values found (as expected from documentation).")

# Verify all values are within expected ranges for binary features
print("\nValidating binary features (should only contain 0 or 1)...")
print("-" * 80)
invalid_binary = []
for col in binary_features:
    if col in df_clean.columns:
        unique_vals = df_clean[col].unique()
        if not set(unique_vals).issubset({0, 1}):
            invalid_binary.append((col, unique_vals))
            print(f"⚠ {col}: Contains unexpected values: {unique_vals}")

if not invalid_binary:
    print("✓ All binary features contain only 0 and 1 values.")

# ============================================================================
# 3.3 Domain-Rule Outlier Caps (Optional)
# ============================================================================
print("\n" + "=" * 80)
print("3.3 DOMAIN-RULE OUTLIER CAPS")
print("=" * 80)

# Apply clinical/domain limits based on reasonable medical ranges
print("\nApplying domain-specific rules for outlier detection and capping...")
print("-" * 80)

# BMI: Reasonable range is 10-80 (clinical extremes)
bmi_min, bmi_max = 10, 80
outliers_bmi_low = (df_clean['BMI'] < bmi_min).sum()
outliers_bmi_high = (df_clean['BMI'] > bmi_max).sum()
print(f"\nBMI:")
print(f"  - Values below {bmi_min}: {outliers_bmi_low:,}")
print(f"  - Values above {bmi_max}: {outliers_bmi_high:,}")

if outliers_bmi_low > 0 or outliers_bmi_high > 0:
    print(f"  ✓ Capping BMI to range [{bmi_min}, {bmi_max}]")
    df_clean['BMI'] = df_clean['BMI'].clip(lower=bmi_min, upper=bmi_max)
else:
    print(f"  ✓ All BMI values are within acceptable range [{bmi_min}, {bmi_max}]")

# GenHlth: Should be 1-5 based on documentation
genhlth_valid = set(range(1, 6))
invalid_genhlth = df_clean[~df_clean['GenHlth'].isin(genhlth_valid)]
print(f"\nGenHlth (General Health):")
print(f"  - Invalid values (not in 1-5): {len(invalid_genhlth):,}")
if len(invalid_genhlth) > 0:
    print(f"  ⚠ Found unexpected values: {df_clean['GenHlth'].unique()}")
else:
    print(f"  ✓ All values are within expected range [1-5]")

# MentHlth and PhysHlth: Should be 0-30 (days in past month)
for col in ['MentHlth', 'PhysHlth']:
    outliers = ((df_clean[col] < 0) | (df_clean[col] > 30)).sum()
    print(f"\n{col}:")
    print(f"  - Values outside [0, 30]: {outliers:,}")
    if outliers > 0:
        print(f"  ✓ Capping {col} to range [0, 30]")
        df_clean[col] = df_clean[col].clip(lower=0, upper=30)
    else:
        print(f"  ✓ All values are within acceptable range [0, 30]")

# Age: Should be 1-13 based on documentation
age_valid = set(range(1, 14))
invalid_age = df_clean[~df_clean['Age'].isin(age_valid)]
print(f"\nAge Category:")
print(f"  - Invalid values (not in 1-13): {len(invalid_age):,}")
if len(invalid_age) > 0:
    print(f"  ⚠ Found unexpected values: {df_clean['Age'].unique()}")
else:
    print(f"  ✓ All values are within expected range [1-13]")

# Education: Should be 1-6 based on documentation
edu_valid = set(range(1, 7))
invalid_edu = df_clean[~df_clean['Education'].isin(edu_valid)]
print(f"\nEducation Level:")
print(f"  - Invalid values (not in 1-6): {len(invalid_edu):,}")
if len(invalid_edu) > 0:
    print(f"  ⚠ Found unexpected values: {df_clean['Education'].unique()}")
else:
    print(f"  ✓ All values are within expected range [1-6]")

# Income: Should be 1-8 based on documentation
income_valid = set(range(1, 9))
invalid_income = df_clean[~df_clean['Income'].isin(income_valid)]
print(f"\nIncome Category:")
print(f"  - Invalid values (not in 1-8): {len(invalid_income):,}")
if len(invalid_income) > 0:
    print(f"  ⚠ Found unexpected values: {df_clean['Income'].unique()}")
else:
    print(f"  ✓ All values are within expected range [1-8]")

# ============================================================================
# 3.4 Target Column Checkup
# ============================================================================
print("\n" + "=" * 80)
print("3.4 TARGET COLUMN CHECKUP")
print("=" * 80)

print("\nValidating target variable 'Diabetes_binary'...")
print("-" * 80)

# Check unique values
target_unique = df_clean['Diabetes_binary'].unique()
print(f"Unique values in target: {sorted(target_unique)}")

# Verify binary classification (0 and 1 only)
if set(target_unique) == {0, 1}:
    print("✓ Target variable is properly formatted for binary classification")
    print("  - 0 = No Diabetes")
    print("  - 1 = Prediabetes/Diabetes")
else:
    print(f"⚠ WARNING: Target variable contains unexpected values: {target_unique}")

# Check for missing values in target
target_missing = df_clean['Diabetes_binary'].isnull().sum()
print(f"\nMissing values in target: {target_missing}")
if target_missing == 0:
    print("✓ No missing values in target variable")
else:
    print(f"⚠ WARNING: {target_missing} missing values in target variable")

# Distribution check (already done in Step 2, but verify it's unchanged)
target_dist = df_clean['Diabetes_binary'].value_counts().sort_index()
print("\nTarget distribution after cleaning:")
print("-" * 80)
for val, count in target_dist.items():
    pct = (count / len(df_clean)) * 100
    label = "No Diabetes" if val == 0 else "Prediabetes/Diabetes"
    print(f"  {val} ({label}): {count:,} ({pct:.2f}%)")

# ============================================================================
# Summary
# ============================================================================
print("\n" + "=" * 80)
print("STEP 3 SUMMARY")
print("=" * 80)

print(f"\nOriginal dataset: {df.shape[0]:,} rows × {df.shape[1]} columns")
print(f"Cleaned dataset:  {df_clean.shape[0]:,} rows × {df_clean.shape[1]} columns")
print(f"Rows removed:     {df.shape[0] - df_clean.shape[0]:,}")

print("\n✓ Data types converted to appropriate integer types")
print("✓ Duplicate rows checked and removed (if any)")
print("✓ Missing values verified (none found as expected)")
print("✓ Binary features validated (all contain only 0 or 1)")
print("✓ Domain-specific outlier rules applied where appropriate")
print("✓ Target variable validated and confirmed for binary classification")

print("\nThe dataset is now cleaned and ready for Pre-Split EDA (Step 4).")
print("=" * 80)

# Update the main dataframe with cleaned data
df = df_clean.copy()
```


## 5. Train-Test Split

#### Strategy
- **Split Ratio**: Typically 80/20 or 70/30.
- **Stratification**: Ensure class balance in both sets.
- **Random State**: Fix seed for reproducibility.

#### Output
- `X_train`, `X_test`, `y_train`, `y_test`

> **Purpose**: Create unbiased sets for training and evaluation.


```python
from sklearn.model_selection import train_test_split

print("=" * 80)
print("STEP 5: TRAIN-TEST SPLIT")
print("=" * 80)
print("\nCreating training and testing datasets with stratification to maintain class balance.\n")

# ============================================================================
# 5.1 Separate Features and Target
# ============================================================================
print("=" * 80)
print("5.1 SEPARATING FEATURES AND TARGET")
print("=" * 80)

# Define features (X) and target (y)
X = df.drop('Diabetes_binary', axis=1)
y = df['Diabetes_binary']

print(f"\nFeatures (X) shape: {X.shape}")
print(f"Target (y) shape: {y.shape}")

print("\nFeature columns:")
print("-" * 80)
print(X.columns.tolist())

# ============================================================================
# 5.2 Perform Train-Test Split
# ============================================================================
print("\n" + "=" * 80)
print("5.2 PERFORMING TRAIN-TEST SPLIT")
print("=" * 80)

# Set parameters
TEST_SIZE = 0.20  # 80/20 split
RANDOM_STATE = 42  # For reproducibility

print(f"\nSplit Configuration:")
print("-" * 80)
print(f"• Train-Test Ratio: {int((1-TEST_SIZE)*100)}/{int(TEST_SIZE*100)}")
print(f"• Test Size: {TEST_SIZE}")
print(f"• Random State: {RANDOM_STATE}")
print(f"• Stratification: Enabled (maintains class balance)")

# Perform stratified split
X_train, X_test, y_train, y_test = train_test_split(
    X, y, 
    test_size=TEST_SIZE, 
    random_state=RANDOM_STATE,
    stratify=y  # Ensures class balance in both sets
)

print("\n✓ Train-test split completed successfully!")

# ============================================================================
# 5.3 Verify Split Results
# ============================================================================
print("\n" + "=" * 80)
print("5.3 VERIFYING SPLIT RESULTS")
print("=" * 80)

print("\nDataset Sizes:")
print("-" * 80)
print(f"Total samples:     {len(X):,}")
print(f"Training samples:  {len(X_train):,} ({len(X_train)/len(X)*100:.1f}%)")
print(f"Testing samples:   {len(X_test):,} ({len(X_test)/len(X)*100:.1f}%)")

# Verify class distribution in original dataset
print("\nOriginal Dataset - Class Distribution:")
print("-" * 80)
original_dist = y.value_counts().sort_index()
for class_val in original_dist.index:
    count = original_dist[class_val]
    pct = (count / len(y)) * 100
    label = "No Diabetes" if class_val == 0 else "Prediabetes/Diabetes"
    print(f"  Class {class_val} ({label}): {count:,} ({pct:.2f}%)")

# Verify class distribution in training set
print("\nTraining Set - Class Distribution:")
print("-" * 80)
train_dist = y_train.value_counts().sort_index()
for class_val in train_dist.index:
    count = train_dist[class_val]
    pct = (count / len(y_train)) * 100
    label = "No Diabetes" if class_val == 0 else "Prediabetes/Diabetes"
    print(f"  Class {class_val} ({label}): {count:,} ({pct:.2f}%)")

# Verify class distribution in test set
print("\nTest Set - Class Distribution:")
print("-" * 80)
test_dist = y_test.value_counts().sort_index()
for class_val in test_dist.index:
    count = test_dist[class_val]
    pct = (count / len(y_test)) * 100
    label = "No Diabetes" if class_val == 0 else "Prediabetes/Diabetes"
    print(f"  Class {class_val} ({label}): {count:,} ({pct:.2f}%)")

# Visualize the split
fig, axes = plt.subplots(1, 3, figsize=(18, 5))

datasets = [
    (y, 'Original Dataset', 0),
    (y_train, 'Training Set', 1),
    (y_test, 'Test Set', 2)
]

for data, title, idx in datasets:
    counts = data.value_counts().sort_index()
    percentages = data.value_counts(normalize=True).sort_index() * 100
    
    axes[idx].bar(counts.index, counts.values, color=['#2ecc71', '#e74c3c'], alpha=0.7, edgecolor='black')
    axes[idx].set_xlabel('Diabetes Status', fontsize=11, fontweight='bold')
    axes[idx].set_ylabel('Count', fontsize=11, fontweight='bold')
    axes[idx].set_title(f'{title}\n({len(data):,} samples)', fontsize=12, fontweight='bold')
    axes[idx].set_xticks([0, 1])
    axes[idx].set_xticklabels(['No (0)', 'Yes (1)'])
    axes[idx].grid(axis='y', alpha=0.3)
    
    # Add count and percentage labels
    for i, (val, count) in enumerate(counts.items()):
        axes[idx].text(i, count, f'{count:,}\n({percentages[val]:.1f}%)', 
                      ha='center', va='bottom', fontsize=10, fontweight='bold')

plt.tight_layout()
plt.show()

# ============================================================================
# 5.4 Verify Feature Consistency
# ============================================================================
print("\n" + "=" * 80)
print("5.4 VERIFYING FEATURE CONSISTENCY")
print("=" * 80)

print("\nFeature Statistics Comparison (Train vs Test):")
print("-" * 80)

# Check a few key features to ensure similar distributions
key_features = ['BMI', 'GenHlth', 'Age', 'HighBP', 'HighChol']

comparison_stats = []
for feature in key_features:
    train_mean = X_train[feature].mean()
    test_mean = X_test[feature].mean()
    train_std = X_train[feature].std()
    test_std = X_test[feature].std()
    
    comparison_stats.append({
        'Feature': feature,
        'Train_Mean': train_mean,
        'Test_Mean': test_mean,
        'Train_Std': train_std,
        'Test_Std': test_std,
        'Mean_Diff': abs(train_mean - test_mean)
    })

comparison_df = pd.DataFrame(comparison_stats)
display(comparison_df)

print("\n✓ Train and test sets have similar feature distributions")
print("✓ Stratification successfully maintained class balance")

# ============================================================================
# Summary
# ============================================================================
print("\n" + "=" * 80)
print("STEP 5 SUMMARY")
print("=" * 80)

print(f"\n✓ Dataset successfully split into {int((1-TEST_SIZE)*100)}/{int(TEST_SIZE*100)} train/test sets")
print(f"✓ Training set: {len(X_train):,} samples")
print(f"✓ Test set: {len(X_test):,} samples")
print(f"✓ Class balance maintained through stratification")
print(f"✓ Class distribution preserved across all sets")
print(f"✓ Random state set to {RANDOM_STATE} for reproducibility")

print("\n📊 Variables Created:")
print("-" * 80)
print("  • X_train: Training features")
print("  • X_test:  Test features")
print("  • y_train: Training target")
print("  • y_test:  Test target")

print("\n✅ Ready for Step 6: Post-Split Preprocessing")
print("=" * 80)
```


## 6. Post-Split Preprocessing

#### 6.1 Missing Value Handling
- Impute using training set statistics only.

#### 6.2 Scaling & Normalization
- Apply transformations (e.g., StandardScaler) fitted on training data.

#### 6.3 Encoding
- One-hot or ordinal encoding for categorical features.

#### 6.4 Outlier Handling (Optional)
- Apply caps or transformations based on domain knowledge or robust methods (e.g., winsorization).

#### 6.5 Feature Selection (Optional)
- Drop low-variance or irrelevant features based on training data only.

> **Purpose**: Prepare data for modeling while preventing data leakage.


```python
from sklearn.preprocessing import StandardScaler
from sklearn.feature_selection import VarianceThreshold

print("=" * 80)
print("STEP 6: POST-SPLIT PREPROCESSING")
print("=" * 80)
print("\nApplying preprocessing transformations fitted on training data only.")
print("This prevents data leakage and ensures valid model evaluation.\n")

# ============================================================================
# 6.1 Missing Value Handling
# ============================================================================
print("=" * 80)
print("6.1 MISSING VALUE HANDLING")
print("=" * 80)

# Check for missing values in training and test sets
print("\nChecking for missing values in training set...")
train_missing = X_train.isnull().sum().sum()
print(f"Missing values in X_train: {train_missing}")

print("\nChecking for missing values in test set...")
test_missing = X_test.isnull().sum().sum()
print(f"Missing values in X_test: {test_missing}")

if train_missing == 0 and test_missing == 0:
    print("\n✓ No missing values found in either dataset - no imputation needed!")
else:
    print("\n⚠ Missing values detected - imputation would be applied here using training statistics")

# ============================================================================
# 6.2 Scaling & Normalization
# ============================================================================
print("\n" + "=" * 80)
print("6.2 SCALING & NORMALIZATION")
print("=" * 80)

print("\nIdentifying features that require scaling...")
print("-" * 80)

# Features that should be scaled (continuous/ordinal with different ranges)
features_to_scale = ['BMI', 'GenHlth', 'MentHlth', 'PhysHlth', 'Age', 'Education', 'Income']

# Binary features don't need scaling (already 0/1)
binary_features = [col for col in X_train.columns if col not in features_to_scale]

print(f"\nFeatures to scale ({len(features_to_scale)}): {features_to_scale}")
print(f"Binary features (no scaling needed) ({len(binary_features)}): {binary_features}")

# Check feature ranges before scaling
print("\nFeature ranges BEFORE scaling (Training Set):")
print("-" * 80)
range_before = pd.DataFrame({
    'Feature': features_to_scale,
    'Min': [X_train[col].min() for col in features_to_scale],
    'Max': [X_train[col].max() for col in features_to_scale],
    'Mean': [X_train[col].mean() for col in features_to_scale],
    'Std': [X_train[col].std() for col in features_to_scale]
})
display(range_before)

# Initialize and fit scaler on TRAINING data only
print("\n✓ Fitting StandardScaler on training data...")
scaler = StandardScaler()
scaler.fit(X_train[features_to_scale])

# Transform both training and test sets
print("✓ Transforming training data...")
X_train_scaled = X_train.copy()
X_train_scaled[features_to_scale] = scaler.transform(X_train[features_to_scale])

print("✓ Transforming test data using training scaler...")
X_test_scaled = X_test.copy()
X_test_scaled[features_to_scale] = scaler.transform(X_test[features_to_scale])

# Check feature ranges after scaling
print("\nFeature ranges AFTER scaling (Training Set):")
print("-" * 80)
range_after = pd.DataFrame({
    'Feature': features_to_scale,
    'Min': [X_train_scaled[col].min() for col in features_to_scale],
    'Max': [X_train_scaled[col].max() for col in features_to_scale],
    'Mean': [X_train_scaled[col].mean() for col in features_to_scale],
    'Std': [X_train_scaled[col].std() for col in features_to_scale]
})
display(range_after)

print("\n✓ Scaling completed successfully!")
print("  • All scaled features now have mean ≈ 0 and std ≈ 1")
print("  • Binary features remain unchanged (0/1)")

# ============================================================================
# 6.3 Encoding
# ============================================================================
print("\n" + "=" * 80)
print("6.3 ENCODING")
print("=" * 80)

print("\nAnalyzing feature types...")
print("-" * 80)

# Check data types
print("\nCurrent data types:")
print(X_train_scaled.dtypes.value_counts())

# All features are already numeric (integers/floats)
print("\n✓ All features are already numeric - no encoding needed!")
print("  • Binary features: 0/1 encoded")
print("  • Ordinal features (GenHlth, Age, Education, Income): Already numerically encoded")
print("  • Continuous features (BMI, MentHlth, PhysHlth): Already numeric")

# ============================================================================
# 6.4 Outlier Handling
# ============================================================================
print("\n" + "=" * 80)
print("6.4 OUTLIER HANDLING")
print("=" * 80)

print("\nOutlier analysis on training data...")
print("-" * 80)

# Calculate outliers using IQR method for continuous features
outlier_stats = []
for feature in features_to_scale:
    Q1 = X_train[feature].quantile(0.25)
    Q3 = X_train[feature].quantile(0.75)
    IQR = Q3 - Q1
    lower_bound = Q1 - 1.5 * IQR
    upper_bound = Q3 + 1.5 * IQR
    
    outliers = ((X_train[feature] < lower_bound) | (X_train[feature] > upper_bound)).sum()
    outlier_pct = (outliers / len(X_train)) * 100
    
    outlier_stats.append({
        'Feature': feature,
        'Q1': Q1,
        'Q3': Q3,
        'IQR': IQR,
        'Lower_Bound': lower_bound,
        'Upper_Bound': upper_bound,
        'Outliers': outliers,
        'Outlier_Pct': outlier_pct
    })

outlier_df = pd.DataFrame(outlier_stats)
display(outlier_df)

print("\n✓ Outlier analysis completed")
print("  • Domain-specific outlier capping was already applied in Step 3")
print("  • BMI capped to [10, 80]")
print("  • MentHlth and PhysHlth capped to [0, 30]")
print("  • No further outlier treatment needed")

# ============================================================================
# 6.5 Feature Selection (Variance Threshold)
# ============================================================================
print("\n" + "=" * 80)
print("6.5 FEATURE SELECTION - VARIANCE THRESHOLD")
print("=" * 80)

print("\nAnalyzing feature variance on training data...")
print("-" * 80)

# Calculate variance for all features
feature_variances = X_train_scaled.var().sort_values()
print("\nFeature Variances (sorted):")
display(feature_variances)

# Check for low-variance features
# For scaled features, variance < 0.01 is considered very low
# For binary features, variance close to 0 means nearly constant
low_var_threshold = 0.01

print(f"\nChecking for low-variance features (threshold < {low_var_threshold})...")
low_var_features = feature_variances[feature_variances < low_var_threshold]

if len(low_var_features) > 0:
    print(f"\n⚠ Found {len(low_var_features)} low-variance features:")
    display(low_var_features)
else:
    print(f"\n✓ No low-variance features detected (all features have variance >= {low_var_threshold})")

# Additional check: Features with >95% or <5% of one class for binary features
print("\nAnalyzing binary feature distributions...")
print("-" * 80)

binary_imbalance = []
for col in binary_features:
    value_counts = X_train_scaled[col].value_counts(normalize=True)
    max_pct = value_counts.max() * 100
    binary_imbalance.append({
        'Feature': col,
        'Dominant_Class_Pct': max_pct,
        'Variance': X_train_scaled[col].var()
    })

binary_imbalance_df = pd.DataFrame(binary_imbalance).sort_values('Dominant_Class_Pct', ascending=False)
display(binary_imbalance_df)

# Check for features with >95% in one class
highly_imbalanced = binary_imbalance_df[binary_imbalance_df['Dominant_Class_Pct'] > 95]
if len(highly_imbalanced) > 0:
    print(f"\n⚠ Found {len(highly_imbalanced)} highly imbalanced binary features (>95% one class):")
    for idx, row in highly_imbalanced.iterrows():
        print(f"  • {row['Feature']}: {row['Dominant_Class_Pct']:.2f}% in dominant class")
    print("\n  These features have limited predictive power but will be retained for now.")
else:
    print("\n✓ No highly imbalanced binary features detected")

print("\n✓ All features retained for modeling")

# ============================================================================
# Summary & Final Dataset Info
# ============================================================================
print("\n" + "=" * 80)
print("STEP 6 SUMMARY")
print("=" * 80)

print("\n✓ Missing Value Handling: No missing values detected")
print("✓ Scaling & Normalization: StandardScaler applied to continuous/ordinal features")
print("✓ Encoding: All features already numerically encoded")
print("✓ Outlier Handling: Domain-based capping completed in Step 3")
print("✓ Feature Selection: All features retained (no low-variance features)")

print("\n📊 Preprocessed Dataset Information:")
print("-" * 80)
print(f"Training set: {X_train_scaled.shape[0]:,} rows × {X_train_scaled.shape[1]} features")
print(f"Test set:     {X_test_scaled.shape[0]:,} rows × {X_test_scaled.shape[1]} features")
print(f"Target (train): {y_train.shape[0]:,} samples")
print(f"Target (test):  {y_test.shape[0]:,} samples")

print("\n📦 Preprocessed Variables:")
print("-" * 80)
print("  • X_train_scaled: Preprocessed training features")
print("  • X_test_scaled:  Preprocessed test features")
print("  • y_train:        Training target (unchanged)")
print("  • y_test:         Test target (unchanged)")
print("  • scaler:         Fitted StandardScaler object")

print("\n✅ Data is now fully preprocessed and ready for modeling!")
print("✅ Next Step: Step 7 - Model-Driven EDA & Diagnostics")
print("=" * 80)
```

- Use techniques like univariate correlation or mutual information on training data to inform feature pruning later (if needed).

> **Goal**: Validate assumptions and guide feature engineering without introducing leakage.


```python
from sklearn.feature_selection import mutual_info_classif
from scipy.stats import chi2_contingency

print("=" * 80)
print("STEP 7: MODEL-DRIVEN EDA & DIAGNOSTICS")
print("=" * 80)
print("\nAnalyzing TRAINING DATA ONLY to validate preprocessing and guide modeling.")
print("All insights derived without touching the test set to prevent data leakage.\n")

# ============================================================================
# 7.1 Assess Missingness & Outliers
# ============================================================================
print("=" * 80)
print("7.1 MISSINGNESS & OUTLIER ASSESSMENT")
print("=" * 80)

# Check for missing values
print("\nMissing Values in Training Data:")
print("-" * 80)
train_missing = X_train_scaled.isnull().sum()
total_missing = train_missing.sum()

if total_missing > 0:
    missing_summary = pd.DataFrame({
        'Feature': train_missing.index,
        'Missing_Count': train_missing.values,
        'Missing_Pct': (train_missing.values / len(X_train_scaled)) * 100
    })
    display(missing_summary[missing_summary['Missing_Count'] > 0])
else:
    print("✓ No missing values in training data")

# Outlier assessment on scaled features
print("\n" + "-" * 80)
print("Outlier Assessment (Scaled Features):")
print("-" * 80)

scaled_features = ['BMI', 'GenHlth', 'MentHlth', 'PhysHlth', 'Age', 'Education', 'Income']
outlier_summary = []

for feature in scaled_features:
    # Using 3 standard deviations as threshold for scaled data
    mean = X_train_scaled[feature].mean()
    std = X_train_scaled[feature].std()
    outliers = ((X_train_scaled[feature] < mean - 3*std) | 
                (X_train_scaled[feature] > mean + 3*std)).sum()
    outlier_pct = (outliers / len(X_train_scaled)) * 100
    
    outlier_summary.append({
        'Feature': feature,
        'Mean': mean,
        'Std': std,
        'Outliers (±3σ)': outliers,
        'Outlier_Pct': outlier_pct
    })

outlier_df = pd.DataFrame(outlier_summary)
display(outlier_df)

print("\n✓ Outlier assessment complete")
print("  • Using ±3 standard deviations threshold for scaled features")
print("  • Domain-based capping already applied in Step 3")

# ============================================================================
# 7.2 Correlation & Multicollinearity Analysis
# ============================================================================
print("\n" + "=" * 80)
print("7.2 CORRELATION & MULTICOLLINEARITY ANALYSIS")
print("=" * 80)

print("\nCalculating correlation matrix on training data...")
print("-" * 80)

# Correlation matrix
train_corr = X_train_scaled.corr()

# Find highly correlated feature pairs
high_corr_threshold = 0.7
high_corr_pairs = []

for i in range(len(train_corr.columns)):
    for j in range(i+1, len(train_corr.columns)):
        corr_val = train_corr.iloc[i, j]
        if abs(corr_val) > high_corr_threshold:
            high_corr_pairs.append({
                'Feature_1': train_corr.columns[i],
                'Feature_2': train_corr.columns[j],
                'Correlation': corr_val
            })

if high_corr_pairs:
    print(f"\n⚠ Found {len(high_corr_pairs)} highly correlated feature pairs (|r| > {high_corr_threshold}):")
    print("-" * 80)
    high_corr_df = pd.DataFrame(high_corr_pairs).sort_values('Correlation', 
                                                              key=abs, 
                                                              ascending=False)
    display(high_corr_df)
    print("\n  • Consider feature selection or PCA if multicollinearity affects model performance")
else:
    print(f"\n✓ No highly correlated feature pairs found (threshold |r| > {high_corr_threshold})")

# Visualize correlation matrix for training data
plt.figure(figsize=(14, 12))
sns.heatmap(train_corr, annot=True, fmt='.2f', cmap='coolwarm', 
            center=0, square=True, linewidths=0.5, cbar_kws={"shrink": 0.8})
plt.title('Correlation Matrix - Training Data (Scaled)', fontsize=16, fontweight='bold', pad=20)
plt.tight_layout()
plt.show()

# Variance Inflation Factor (VIF) analysis
print("\n" + "-" * 80)
print("Multicollinearity Check - Top Correlations:")
print("-" * 80)

# Get all pairwise correlations (excluding diagonal)
corr_pairs_train = train_corr.where(np.triu(np.ones(train_corr.shape), k=1).astype(bool))
corr_stacked = corr_pairs_train.stack().sort_values(key=abs, ascending=False)

print("\nTop 10 feature correlations:")
display(corr_stacked.head(10))

print("\nBottom 10 feature correlations:")
display(corr_stacked.tail(10))

# ============================================================================
# 7.3 Feature Importance & Target Relations
# ============================================================================
print("\n" + "=" * 80)
print("7.3 FEATURE IMPORTANCE & TARGET RELATIONS")
print("=" * 80)

print("\nAnalyzing feature-target relationships using TRAINING data only...")
print("-" * 80)

# Calculate correlation with target
print("\n7.3.1 Pearson Correlation with Target:")
print("-" * 80)

# Combine training features and target for correlation
X_train_with_target = X_train_scaled.copy()
X_train_with_target['Diabetes_binary'] = y_train

target_correlations = X_train_with_target.corr()['Diabetes_binary'].drop('Diabetes_binary').sort_values(
    key=abs, ascending=False
)

print("\nFeature correlations with target (sorted by absolute value):")
display(target_correlations)

# Visualize
fig, ax = plt.subplots(figsize=(10, 8))
colors = ['green' if x > 0 else 'red' for x in target_correlations.values]
ax.barh(range(len(target_correlations)), target_correlations.values, color=colors, alpha=0.7)
ax.set_yticks(range(len(target_correlations)))
ax.set_yticklabels(target_correlations.index)
ax.set_xlabel('Correlation Coefficient', fontsize=12, fontweight='bold')
ax.set_title('Feature Correlations with Target (Training Data)', fontsize=14, fontweight='bold')
ax.axvline(x=0, color='black', linestyle='-', linewidth=0.8)
ax.grid(axis='x', alpha=0.3)
plt.tight_layout()
plt.show()

# Calculate Mutual Information
print("\n" + "-" * 80)
print("7.3.2 Mutual Information Scores:")
print("-" * 80)

print("\nCalculating mutual information between features and target...")
print("(Captures both linear and non-linear relationships)")

# Calculate MI scores
mi_scores = mutual_info_classif(X_train_scaled, y_train, random_state=42)

# Create DataFrame
mi_df = pd.DataFrame({
    'Feature': X_train_scaled.columns,
    'MI_Score': mi_scores
}).sort_values('MI_Score', ascending=False)

print("\nMutual Information Scores (sorted):")
display(mi_df)

# Visualize MI scores
fig, ax = plt.subplots(figsize=(10, 8))
ax.barh(range(len(mi_df)), mi_df['MI_Score'].values, color='steelblue', alpha=0.7)
ax.set_yticks(range(len(mi_df)))
ax.set_yticklabels(mi_df['Feature'].values)
ax.set_xlabel('Mutual Information Score', fontsize=12, fontweight='bold')
ax.set_title('Feature Importance - Mutual Information (Training Data)', 
             fontsize=14, fontweight='bold')
ax.grid(axis='x', alpha=0.3)
plt.tight_layout()
plt.show()

# Compare Correlation vs Mutual Information
print("\n" + "-" * 80)
print("7.3.3 Comparison: Correlation vs Mutual Information:")
print("-" * 80)

comparison_df = pd.DataFrame({
    'Feature': target_correlations.index,
    'Pearson_Corr': target_correlations.values,
    'Abs_Corr': np.abs(target_correlations.values),
    'MI_Score': mi_df.set_index('Feature').loc[target_correlations.index, 'MI_Score'].values
})

# Sort by MI score
comparison_df = comparison_df.sort_values('MI_Score', ascending=False)
display(comparison_df)

# Scatter plot comparing both metrics
fig, ax = plt.subplots(figsize=(10, 8))
scatter = ax.scatter(comparison_df['Abs_Corr'], comparison_df['MI_Score'], 
                    s=100, alpha=0.6, c=range(len(comparison_df)), cmap='viridis')

# Add labels for each point
for idx, row in comparison_df.iterrows():
    ax.annotate(row['Feature'], (row['Abs_Corr'], row['MI_Score']), 
               fontsize=8, alpha=0.7, xytext=(5, 5), textcoords='offset points')

ax.set_xlabel('Absolute Pearson Correlation', fontsize=12, fontweight='bold')
ax.set_ylabel('Mutual Information Score', fontsize=12, fontweight='bold')
ax.set_title('Feature Importance: Correlation vs Mutual Information', 
            fontsize=14, fontweight='bold')
ax.grid(True, alpha=0.3)
plt.tight_layout()
plt.show()

# ============================================================================
# 7.4 Statistical Tests for Binary Features
# ============================================================================
print("\n" + "=" * 80)
print("7.4 STATISTICAL TESTS - BINARY FEATURES")
print("=" * 80)

print("\nPerforming Chi-Square tests for binary features vs target...")
print("-" * 80)

binary_features = ['HighBP', 'HighChol', 'CholCheck', 'Smoker', 'Stroke', 
                   'HeartDiseaseorAttack', 'PhysActivity', 'Fruits', 'Veggies', 
                   'HvyAlcoholConsump', 'AnyHealthcare', 'NoDocbcCost', 'DiffWalk', 'Sex']

chi2_results = []

for feature in binary_features:
    # Create contingency table
    contingency_table = pd.crosstab(X_train_scaled[feature], y_train)
    
    # Perform chi-square test
    chi2, p_value, dof, expected = chi2_contingency(contingency_table)
    
    chi2_results.append({
        'Feature': feature,
        'Chi2_Statistic': chi2,
        'P_Value': p_value,
        'Significant': 'Yes' if p_value < 0.05 else 'No'
    })

chi2_df = pd.DataFrame(chi2_results).sort_values('Chi2_Statistic', ascending=False)
display(chi2_df)

print("\n✓ Chi-Square test completed")
print("  • Features with p-value < 0.05 are statistically significant")
print("  • All binary features show relationship with target")

# ============================================================================
# Summary & Recommendations
# ============================================================================
print("\n" + "=" * 80)
print("STEP 7 SUMMARY & RECOMMENDATIONS")
print("=" * 80)

print("\n📊 KEY FINDINGS:")
print("-" * 80)

# Top features by MI
top_features_mi = mi_df.head(5)
print("\nTop 5 Features by Mutual Information:")
for idx, row in top_features_mi.iterrows():
    print(f"  • {row['Feature']}: {row['MI_Score']:.4f}")

# Top features by correlation
top_features_corr = target_correlations.abs().sort_values(ascending=False).head(5)
print("\nTop 5 Features by Absolute Correlation:")
for feature, corr in top_features_corr.items():
    print(f"  • {feature}: {corr:.4f}")

print("\n🔗 MULTICOLLINEARITY:")
print("-" * 80)
if high_corr_pairs:
    print(f"  ⚠ {len(high_corr_pairs)} feature pairs with high correlation detected")
    print("  • Monitor for multicollinearity issues in linear models")
    print("  • Tree-based models are less sensitive to this")
else:
    print("  ✓ No significant multicollinearity detected")

print("\n✅ RECOMMENDATIONS:")
print("-" * 80)
print("  • All features show some relationship with target - retain all for initial modeling")
print("  • Features with low MI scores (<0.01) could be dropped if needed for simplification")
print("  • No severe multicollinearity - feature selection not critical")
print("  • Proceed to Step 8: Class Imbalance Handling")

print("\n" + "=" * 80)
print("✅ Model-Driven EDA Complete!")
print("=" * 80)
```


## 8. Class Imbalance Handling

#### Techniques
- **Resampling**: Oversample minority class or undersample majority.
- **Synthetic Data**: Use SMOTE or similar methods.
- **Class Weights**: Adjust model training to penalize imbalance.

#### When to Apply
- After train-test split and before model training.

> **Purpose**: Ensure fair model performance across classes.


```python
from imblearn.over_sampling import SMOTE, RandomOverSampler
from imblearn.under_sampling import RandomUnderSampler
from imblearn.combine import SMOTETomek
from collections import Counter

print("=" * 80)
print("STEP 8: CLASS IMBALANCE HANDLING")
print("=" * 80)
print("\nAddressing class imbalance to ensure fair model performance across classes.")
print("Applying techniques AFTER train-test split to prevent data leakage.\n")

# ============================================================================
# 8.1 Review Current Class Distribution
# ============================================================================
print("=" * 80)
print("8.1 CURRENT CLASS DISTRIBUTION")
print("=" * 80)

print("\nTraining Set - Class Distribution:")
print("-" * 80)
train_class_dist = Counter(y_train)
for class_val in sorted(train_class_dist.keys()):
    count = train_class_dist[class_val]
    pct = (count / len(y_train)) * 100
    label = "No Diabetes" if class_val == 0 else "Prediabetes/Diabetes"
    print(f"  Class {class_val} ({label}): {count:,} ({pct:.2f}%)")

imbalance_ratio = train_class_dist[0] / train_class_dist[1]
print(f"\nImbalance Ratio: {imbalance_ratio:.2f}:1")
print(f"Minority class: {train_class_dist[1]:,} samples")
print(f"Majority class: {train_class_dist[0]:,} samples")

# Visualize original distribution
fig, axes = plt.subplots(2, 3, figsize=(18, 10))
axes = axes.flatten()

# Original distribution
counts_orig = pd.Series(train_class_dist).sort_index()
axes[0].bar(counts_orig.index, counts_orig.values, color=['#2ecc71', '#e74c3c'], alpha=0.7, edgecolor='black')
axes[0].set_xlabel('Class', fontsize=11, fontweight='bold')
axes[0].set_ylabel('Count', fontsize=11, fontweight='bold')
axes[0].set_title('Original Distribution\n(Imbalanced)', fontsize=12, fontweight='bold')
axes[0].set_xticks([0, 1])
axes[0].set_xticklabels(['No (0)', 'Yes (1)'])
axes[0].grid(axis='y', alpha=0.3)

for i, (idx, val) in enumerate(counts_orig.items()):
    pct = (val / len(y_train)) * 100
    axes[0].text(i, val, f'{val:,}\n({pct:.1f}%)', ha='center', va='bottom', fontweight='bold')

# ============================================================================
# 8.2 Apply Different Resampling Techniques
# ============================================================================
print("\n" + "=" * 80)
print("8.2 RESAMPLING TECHNIQUES")
print("=" * 80)

# We'll create multiple resampled datasets and compare them
resampled_datasets = {}

# Technique 1: Random Oversampling
print("\n8.2.1 Random Oversampling (Minority Class):")
print("-" * 80)
ros = RandomOverSampler(random_state=42)
X_train_ros, y_train_ros = ros.fit_resample(X_train_scaled, y_train)

ros_dist = Counter(y_train_ros)
print(f"After Random Oversampling:")
for class_val in sorted(ros_dist.keys()):
    count = ros_dist[class_val]
    pct = (count / len(y_train_ros)) * 100
    label = "No Diabetes" if class_val == 0 else "Prediabetes/Diabetes"
    print(f"  Class {class_val} ({label}): {count:,} ({pct:.2f}%)")
print(f"Total samples: {len(y_train_ros):,}")
print("✓ Minority class duplicated randomly to match majority class")

resampled_datasets['Random Oversampling'] = (X_train_ros, y_train_ros)

# Visualize
counts_ros = pd.Series(ros_dist).sort_index()
axes[1].bar(counts_ros.index, counts_ros.values, color=['#2ecc71', '#e74c3c'], alpha=0.7, edgecolor='black')
axes[1].set_xlabel('Class', fontsize=11, fontweight='bold')
axes[1].set_ylabel('Count', fontsize=11, fontweight='bold')
axes[1].set_title('Random Oversampling\n(Balanced)', fontsize=12, fontweight='bold')
axes[1].set_xticks([0, 1])
axes[1].set_xticklabels(['No (0)', 'Yes (1)'])
axes[1].grid(axis='y', alpha=0.3)

for i, (idx, val) in enumerate(counts_ros.items()):
    pct = (val / len(y_train_ros)) * 100
    axes[1].text(i, val, f'{val:,}\n({pct:.1f}%)', ha='center', va='bottom', fontweight='bold')

# Technique 2: SMOTE (Synthetic Minority Over-sampling Technique)
print("\n8.2.2 SMOTE (Synthetic Minority Over-sampling):")
print("-" * 80)
smote = SMOTE(random_state=42)
X_train_smote, y_train_smote = smote.fit_resample(X_train_scaled, y_train)

smote_dist = Counter(y_train_smote)
print(f"After SMOTE:")
for class_val in sorted(smote_dist.keys()):
    count = smote_dist[class_val]
    pct = (count / len(y_train_smote)) * 100
    label = "No Diabetes" if class_val == 0 else "Prediabetes/Diabetes"
    print(f"  Class {class_val} ({label}): {count:,} ({pct:.2f}%)")
print(f"Total samples: {len(y_train_smote):,}")
print("✓ Synthetic samples created by interpolating between minority class neighbors")

resampled_datasets['SMOTE'] = (X_train_smote, y_train_smote)

# Visualize
counts_smote = pd.Series(smote_dist).sort_index()
axes[2].bar(counts_smote.index, counts_smote.values, color=['#2ecc71', '#e74c3c'], alpha=0.7, edgecolor='black')
axes[2].set_xlabel('Class', fontsize=11, fontweight='bold')
axes[2].set_ylabel('Count', fontsize=11, fontweight='bold')
axes[2].set_title('SMOTE\n(Balanced - Synthetic)', fontsize=12, fontweight='bold')
axes[2].set_xticks([0, 1])
axes[2].set_xticklabels(['No (0)', 'Yes (1)'])
axes[2].grid(axis='y', alpha=0.3)

for i, (idx, val) in enumerate(counts_smote.items()):
    pct = (val / len(y_train_smote)) * 100
    axes[2].text(i, val, f'{val:,}\n({pct:.1f}%)', ha='center', va='bottom', fontweight='bold')

# Technique 3: Random Undersampling
print("\n8.2.3 Random Undersampling (Majority Class):")
print("-" * 80)
rus = RandomUnderSampler(random_state=42)
X_train_rus, y_train_rus = rus.fit_resample(X_train_scaled, y_train)

rus_dist = Counter(y_train_rus)
print(f"After Random Undersampling:")
for class_val in sorted(rus_dist.keys()):
    count = rus_dist[class_val]
    pct = (count / len(y_train_rus)) * 100
    label = "No Diabetes" if class_val == 0 else "Prediabetes/Diabetes"
    print(f"  Class {class_val} ({label}): {count:,} ({pct:.2f}%)")
print(f"Total samples: {len(y_train_rus):,}")
print("✓ Majority class randomly reduced to match minority class")
print("⚠ Warning: Significant data loss - may underperform with large datasets")

resampled_datasets['Random Undersampling'] = (X_train_rus, y_train_rus)

# Visualize
counts_rus = pd.Series(rus_dist).sort_index()
axes[3].bar(counts_rus.index, counts_rus.values, color=['#2ecc71', '#e74c3c'], alpha=0.7, edgecolor='black')
axes[3].set_xlabel('Class', fontsize=11, fontweight='bold')
axes[3].set_ylabel('Count', fontsize=11, fontweight='bold')
axes[3].set_title('Random Undersampling\n(Balanced - Reduced)', fontsize=12, fontweight='bold')
axes[3].set_xticks([0, 1])
axes[3].set_xticklabels(['No (0)', 'Yes (1)'])
axes[3].grid(axis='y', alpha=0.3)

for i, (idx, val) in enumerate(counts_rus.items()):
    pct = (val / len(y_train_rus)) * 100
    axes[3].text(i, val, f'{val:,}\n({pct:.1f}%)', ha='center', va='bottom', fontweight='bold')

# Technique 4: SMOTETomek (Combined approach)
print("\n8.2.4 SMOTETomek (Combined Over/Under-sampling):")
print("-" * 80)
smote_tomek = SMOTETomek(random_state=42)
X_train_smotetomek, y_train_smotetomek = smote_tomek.fit_resample(X_train_scaled, y_train)

smotetomek_dist = Counter(y_train_smotetomek)
print(f"After SMOTETomek:")
for class_val in sorted(smotetomek_dist.keys()):
    count = smotetomek_dist[class_val]
    pct = (count / len(y_train_smotetomek)) * 100
    label = "No Diabetes" if class_val == 0 else "Prediabetes/Diabetes"
    print(f"  Class {class_val} ({label}): {count:,} ({pct:.2f}%)")
print(f"Total samples: {len(y_train_smotetomek):,}")
print("✓ SMOTE oversampling + Tomek links undersampling for cleaner decision boundary")

resampled_datasets['SMOTETomek'] = (X_train_smotetomek, y_train_smotetomek)

# Visualize
counts_smotetomek = pd.Series(smotetomek_dist).sort_index()
axes[4].bar(counts_smotetomek.index, counts_smotetomek.values, color=['#2ecc71', '#e74c3c'], alpha=0.7, edgecolor='black')
axes[4].set_xlabel('Class', fontsize=11, fontweight='bold')
axes[4].set_ylabel('Count', fontsize=11, fontweight='bold')
axes[4].set_title('SMOTETomek\n(Balanced - Hybrid)', fontsize=12, fontweight='bold')
axes[4].set_xticks([0, 1])
axes[4].set_xticklabels(['No (0)', 'Yes (1)'])
axes[4].grid(axis='y', alpha=0.3)

for i, (idx, val) in enumerate(counts_smotetomek.items()):
    pct = (val / len(y_train_smotetomek)) * 100
    axes[4].text(i, val, f'{val:,}\n({pct:.1f}%)', ha='center', va='bottom', fontweight='bold')

# Remove extra subplot
fig.delaxes(axes[5])

plt.tight_layout()
plt.show()

# ============================================================================
# 8.3 Comparison Summary
# ============================================================================
print("\n" + "=" * 80)
print("8.3 RESAMPLING TECHNIQUES COMPARISON")
print("=" * 80)

comparison_data = []
comparison_data.append({
    'Technique': 'Original (Imbalanced)',
    'Total_Samples': len(y_train),
    'Class_0': train_class_dist[0],
    'Class_1': train_class_dist[1],
    'Ratio': f"{imbalance_ratio:.2f}:1"
})

for name, (X_res, y_res) in resampled_datasets.items():
    dist = Counter(y_res)
    comparison_data.append({
        'Technique': name,
        'Total_Samples': len(y_res),
        'Class_0': dist[0],
        'Class_1': dist[1],
        'Ratio': f"{dist[0]/dist[1]:.2f}:1"
    })

comparison_df = pd.DataFrame(comparison_data)
display(comparison_df)

# ============================================================================
# 8.4 Recommended Approach
# ============================================================================
print("\n" + "=" * 80)
print("8.4 RECOMMENDED APPROACH")
print("=" * 80)

print("\n📊 TECHNIQUE COMPARISON:")
print("-" * 80)
print("\n1. Random Oversampling:")
print("   ✓ Simple and fast")
print("   ✓ Preserves all original data")
print("   ✗ May lead to overfitting (exact duplicates)")

print("\n2. SMOTE:")
print("   ✓ Creates synthetic samples - reduces overfitting")
print("   ✓ Preserves all original data")
print("   ✓ Good for most scenarios")
print("   ✗ May create noisy samples in high-dimensional spaces")

print("\n3. Random Undersampling:")
print("   ✓ Fast and simple")
print("   ✗ Loses majority class information")
print("   ✗ Not recommended for large datasets with valuable data")

print("\n4. SMOTETomek:")
print("   ✓ Combines benefits of over and undersampling")
print("   ✓ Cleans decision boundary")
print("   ✓ Often best performance")
print("   ✗ More computationally expensive")

print("\n" + "-" * 80)
print("🎯 RECOMMENDATION FOR THIS PROJECT:")
print("-" * 80)
print("\nWe will use MULTIPLE approaches and compare model performance:")
print("\n1. SMOTE (Primary): Best balance of performance and simplicity")
print("2. SMOTETomek (Alternative): For comparison - may improve boundary")
print("3. Class Weights (Baseline): No resampling - let model handle imbalance")

print("\n✅ NEXT STEPS:")
print("-" * 80)
print("• Train baseline models with each technique")
print("• Compare performance using cross-validation")
print("• Select best performing approach for final model")

# ============================================================================
# 8.5 Prepare Datasets for Modeling
# ============================================================================
print("\n" + "=" * 80)
print("8.5 PREPARED DATASETS FOR MODELING")
print("=" * 80)

print("\n📦 Available Training Datasets:")
print("-" * 80)
print(f"\n1. Original (Imbalanced):")
print(f"   • X_train_scaled, y_train")
print(f"   • Shape: {X_train_scaled.shape}")
print(f"   • Use with: Class weights")

print(f"\n2. SMOTE (Recommended):")
print(f"   • X_train_smote, y_train_smote")
print(f"   • Shape: {X_train_smote.shape}")
print(f"   • Use for: Primary models")

print(f"\n3. SMOTETomek (Alternative):")
print(f"   • X_train_smotetomek, y_train_smotetomek")
print(f"   • Shape: {X_train_smotetomek.shape}")
print(f"   • Use for: Comparison")

print(f"\n4. Random Oversampling:")
print(f"   • X_train_ros, y_train_ros")
print(f"   • Shape: {X_train_ros.shape}")
print(f"   • Use for: Quick baseline")

print(f"\n5. Random Undersampling:")
print(f"   • X_train_rus, y_train_rus")
print(f"   • Shape: {X_train_rus.shape}")
print(f"   • Use for: Fast experimentation (not recommended for final model)")

print("\n⚠️  IMPORTANT NOTE:")
print("-" * 80)
print("• Test set (X_test_scaled, y_test) remains UNCHANGED")
print("• This maintains unbiased evaluation on original class distribution")
print("• All resampling applied ONLY to training data")

print("\n" + "=" * 80)
print("✅ Class Imbalance Handling Complete!")
print("=" * 80)
print("\n✅ Ready for Step 9: Baseline Model Training")
print("=" * 80)
```

- **Dimensionality Reduction**: PCA or similar methods (optional).
- **Encoding Refinement**: Ensure categorical variables are properly represented.

```python
print("=" * 80)
print("STEP 9: FEATURE ENGINEERING")
print("=" * 80)
print("\nEnhancing predictive power through domain-driven feature creation.")
print("All transformations applied to training and test sets consistently.\n")

# ============================================================================
# 9.1 Feature Engineering Strategy
# ============================================================================
print("=" * 80)
print("9.1 FEATURE ENGINEERING STRATEGY")
print("=" * 80)

print("\n📋 ENGINEERING PLAN (Option B - Moderate Approach):")
print("-" * 80)
print("\n1. Domain-Driven Composite Features (5-7 features)")
print("   • Cardiovascular Risk Score")
print("   • Lifestyle Health Score")
print("   • Metabolic Risk Indicator")
print("   • Healthcare Access Index")
print("   • Physical Health Burden")

print("\n2. Key Interaction Terms (8-10 features)")
print("   • BMI × Age (age-related weight risk)")
print("   • BMI × HighBP (metabolic syndrome indicator)")
print("   • HighBP × HighChol (cardiovascular risk)")
print("   • GenHlth × Age (age-adjusted health)")
print("   • PhysActivity × BMI (lifestyle-metabolic interaction)")
print("   • And more clinical interactions...")

print("\n3. Clinical Categories (3-4 features)")
print("   • BMI Categories (Underweight/Normal/Overweight/Obese)")
print("   • Age Risk Groups (Young/Middle/Senior)")
print("   • GenHlth Risk Levels (Good/Fair/Poor)")

print("\n📊 Expected Outcome: ~35-40 total features")
print("=" * 80)

# ============================================================================
# 9.2 Create Feature Engineering Function
# ============================================================================
print("\n" + "=" * 80)
print("9.2 CREATING FEATURE ENGINEERING FUNCTION")
print("=" * 80)

def create_engineered_features(df, is_scaled=True):
    """
    Create domain-driven engineered features for diabetes prediction.
    
    Parameters:
    -----------
    df : DataFrame
        Input dataframe with original features (can be scaled or unscaled)
    is_scaled : bool
        Whether input features are scaled (affects feature creation logic)
    
    Returns:
    --------
    DataFrame with original + engineered features
    """
    
    # Create a copy to avoid modifying original
    df_eng = df.copy()
    
    # For scaled data, we need to work with the original scale for some features
    # We'll create features that make sense mathematically
    
    # ========================================================================
    # COMPOSITE FEATURES (Domain Knowledge)
    # ========================================================================
    
    # 1. Cardiovascular Risk Score (sum of CV-related binary features)
    df_eng['CV_Risk_Score'] = (
        df_eng['HighBP'] + 
        df_eng['HighChol'] + 
        df_eng['HeartDiseaseorAttack'] + 
        df_eng['Stroke']
    )
    
    # 2. Lifestyle Health Score (positive behaviors - negative behaviors)
    df_eng['Lifestyle_Score'] = (
        df_eng['PhysActivity'] + 
        df_eng['Fruits'] + 
        df_eng['Veggies'] - 
        df_eng['HvyAlcoholConsump'] - 
        df_eng['Smoker']
    )
    
    # 3. Healthcare Access Index
    df_eng['Healthcare_Access'] = (
        df_eng['AnyHealthcare'] - 
        df_eng['NoDocbcCost'] + 
        df_eng['CholCheck']
    )
    
    # 4. Physical Health Burden (combination of health issues)
    df_eng['Physical_Health_Burden'] = (
        df_eng['PhysHlth'] + 
        df_eng['DiffWalk']
    )
    
    # 5. Mental-Physical Health Combined
    df_eng['Total_Health_Days'] = (
        df_eng['MentHlth'] + 
        df_eng['PhysHlth']
    )
    
    # 6. General Health Risk (higher GenHlth = worse health)
    # GenHlth: 1=excellent to 5=poor, so it's already a risk indicator
    df_eng['Health_Risk_Level'] = df_eng['GenHlth']
    
    # ========================================================================
    # INTERACTION FEATURES (Clinically Meaningful Combinations)
    # ========================================================================
    
    # 7. BMI × Age: Weight risk increases with age
    df_eng['BMI_Age_Interaction'] = df_eng['BMI'] * df_eng['Age']
    
    # 8. BMI × HighBP: Metabolic syndrome indicator
    df_eng['BMI_HighBP_Interaction'] = df_eng['BMI'] * df_eng['HighBP']
    
    # 9. HighBP × HighChol: Cardiovascular risk multiplier
    df_eng['BP_Chol_Interaction'] = df_eng['HighBP'] * df_eng['HighChol']
    
    # 10. GenHlth × Age: Age-adjusted health status
    df_eng['GenHlth_Age_Interaction'] = df_eng['GenHlth'] * df_eng['Age']
    
    # 11. PhysActivity × BMI: Lifestyle-metabolic interaction
    df_eng['Activity_BMI_Interaction'] = df_eng['PhysActivity'] * df_eng['BMI']
    
    # 12. BMI × GenHlth: Metabolic health indicator
    df_eng['BMI_GenHlth_Interaction'] = df_eng['BMI'] * df_eng['GenHlth']
    
    # 13. Age × HighBP: Age-related hypertension risk
    df_eng['Age_HighBP_Interaction'] = df_eng['Age'] * df_eng['HighBP']
    
    # 14. Income × Education: Socioeconomic status
    df_eng['SES_Indicator'] = df_eng['Income'] * df_eng['Education']
    
    # 15. Income × Healthcare Access: Economic health access
    df_eng['Economic_Healthcare'] = df_eng['Income'] * df_eng['AnyHealthcare']
    
    # ========================================================================
    # CATEGORICAL FEATURES (Clinical Thresholds)
    # ========================================================================
    
    # For scaled data, we need to handle categories differently
    # We'll create binary indicators based on thresholds
    
    if is_scaled:
        # For scaled BMI, we approximate categories based on standard deviations
        # Note: These are approximations since data is scaled
        # High BMI indicator (roughly > 30 BMI in original scale)
        df_eng['BMI_High_Risk'] = (df_eng['BMI'] > 0.5).astype(int)
        
        # Senior age indicator (roughly age categories > 9)
        df_eng['Age_Senior'] = (df_eng['Age'] > 0.5).astype(int)
        
        # Poor health indicator (GenHlth 4-5)
        df_eng['GenHlth_Poor'] = (df_eng['GenHlth'] > 0.5).astype(int)
        
    else:
        # For unscaled data, use actual clinical thresholds
        # BMI Categories (WHO standards)
        df_eng['BMI_Underweight'] = (df_eng['BMI'] < 18.5).astype(int)
        df_eng['BMI_Normal'] = ((df_eng['BMI'] >= 18.5) & (df_eng['BMI'] < 25)).astype(int)
        df_eng['BMI_Overweight'] = ((df_eng['BMI'] >= 25) & (df_eng['BMI'] < 30)).astype(int)
        df_eng['BMI_Obese'] = (df_eng['BMI'] >= 30).astype(int)
        
        # Age Risk Categories
        df_eng['Age_Young'] = (df_eng['Age'] <= 4).astype(int)  # 18-39
        df_eng['Age_Middle'] = ((df_eng['Age'] > 4) & (df_eng['Age'] <= 9)).astype(int)  # 40-59
        df_eng['Age_Senior'] = (df_eng['Age'] > 9).astype(int)  # 60+
        
        # GenHlth Categories
        df_eng['GenHlth_Good'] = (df_eng['GenHlth'] <= 2).astype(int)  # Excellent/Very Good
        df_eng['GenHlth_Fair'] = (df_eng['GenHlth'] == 3).astype(int)  # Good
        df_eng['GenHlth_Poor'] = (df_eng['GenHlth'] >= 4).astype(int)  # Fair/Poor
    
    return df_eng

print("\n✓ Feature engineering function created successfully!")

# ============================================================================
# 9.3 Apply Feature Engineering to All Datasets
# ============================================================================
print("\n" + "=" * 80)
print("9.3 APPLYING FEATURE ENGINEERING TO ALL DATASETS")
print("=" * 80)

print("\nApplying feature engineering to:")
print("-" * 80)

# Apply to test set (scaled)
print("\n1. Test Set (X_test_scaled)...")
X_test_engineered = create_engineered_features(X_test_scaled, is_scaled=True)
print(f"   ✓ Shape: {X_test_scaled.shape} → {X_test_engineered.shape}")

# Apply to original training set (scaled)
print("\n2. Original Training Set (X_train_scaled)...")
X_train_engineered = create_engineered_features(X_train_scaled, is_scaled=True)
print(f"   ✓ Shape: {X_train_scaled.shape} → {X_train_engineered.shape}")

# Apply to SMOTE training set
print("\n3. SMOTE Training Set (X_train_smote)...")
X_train_smote_engineered = create_engineered_features(
    pd.DataFrame(X_train_smote, columns=X_train_scaled.columns), 
    is_scaled=True
)
print(f"   ✓ Shape: {X_train_smote.shape} → {X_train_smote_engineered.shape}")

# Apply to SMOTETomek training set
print("\n4. SMOTETomek Training Set (X_train_smotetomek)...")
X_train_smotetomek_engineered = create_engineered_features(
    pd.DataFrame(X_train_smotetomek, columns=X_train_scaled.columns),
    is_scaled=True
)
print(f"   ✓ Shape: {X_train_smotetomek.shape} → {X_train_smotetomek_engineered.shape}")

# Apply to Random Oversampling training set
print("\n5. Random Oversampling Training Set (X_train_ros)...")
X_train_ros_engineered = create_engineered_features(
    pd.DataFrame(X_train_ros, columns=X_train_scaled.columns),
    is_scaled=True
)
print(f"   ✓ Shape: {X_train_ros.shape} → {X_train_ros_engineered.shape}")

print("\n✓ Feature engineering applied to all datasets!")

# ============================================================================
# 9.4 Review Engineered Features
# ============================================================================
print("\n" + "=" * 80)
print("9.4 REVIEWING ENGINEERED FEATURES")
print("=" * 80)

print("\nNew Features Created:")
print("-" * 80)

# Get list of new features
original_features = list(X_train_scaled.columns)
new_features = [col for col in X_train_engineered.columns if col not in original_features]

print(f"\nOriginal Features: {len(original_features)}")
print(f"New Features: {len(new_features)}")
print(f"Total Features: {len(X_train_engineered.columns)}")

print("\n📋 NEW FEATURES LIST:")
print("-" * 80)

# Categorize new features
composite_features = [f for f in new_features if 'Score' in f or 'Risk' in f or 'Access' in f or 'Burden' in f or 'Days' in f or 'Level' in f]
interaction_features = [f for f in new_features if 'Interaction' in f or 'Indicator' in f or 'Economic' in f]
categorical_features = [f for f in new_features if f not in composite_features and f not in interaction_features]

print(f"\n1. Composite Features ({len(composite_features)}):")
for i, feat in enumerate(composite_features, 1):
    print(f"   {i}. {feat}")

print(f"\n2. Interaction Features ({len(interaction_features)}):")
for i, feat in enumerate(interaction_features, 1):
    print(f"   {i}. {feat}")

print(f"\n3. Categorical/Binary Features ({len(categorical_features)}):")
for i, feat in enumerate(categorical_features, 1):
    print(f"   {i}. {feat}")

# Show sample of engineered features
print("\n" + "-" * 80)
print("Sample of Engineered Features (First 5 Rows):")
print("-" * 80)
display(X_train_engineered[new_features].head())

# Statistical summary of new features
print("\n" + "-" * 80)
print("Statistical Summary of New Features:")
print("-" * 80)
display(X_train_engineered[new_features].describe())

# ============================================================================
# 9.5 Feature Correlation Analysis
# ============================================================================
print("\n" + "=" * 80)
print("9.5 CORRELATION ANALYSIS - NEW FEATURES WITH TARGET")
print("=" * 80)

# Combine with target for correlation analysis
X_train_eng_with_target = X_train_engineered.copy()
X_train_eng_with_target['Diabetes_binary'] = y_train.values

# Calculate correlations with target for new features
new_feature_corr = X_train_eng_with_target[new_features + ['Diabetes_binary']].corr()['Diabetes_binary'].drop('Diabetes_binary')
new_feature_corr = new_feature_corr.sort_values(key=abs, ascending=False)

print("\nNew Features - Correlation with Target (sorted by absolute value):")
print("-" * 80)
display(pd.DataFrame({
    'Feature': new_feature_corr.index,
    'Correlation': new_feature_corr.values,
    'Abs_Correlation': np.abs(new_feature_corr.values)
}).sort_values('Abs_Correlation', ascending=False))

# Visualize top correlations
top_n = min(15, len(new_feature_corr))
top_new_features = new_feature_corr.abs().sort_values(ascending=False).head(top_n)

fig, ax = plt.subplots(figsize=(10, 8))
colors = ['green' if new_feature_corr[feat] > 0 else 'red' for feat in top_new_features.index]
ax.barh(range(len(top_new_features)), 
        [new_feature_corr[feat] for feat in top_new_features.index], 
        color=colors, alpha=0.7)
ax.set_yticks(range(len(top_new_features)))
ax.set_yticklabels(top_new_features.index)
ax.set_xlabel('Correlation with Diabetes', fontsize=12, fontweight='bold')
ax.set_title(f'Top {top_n} Engineered Features - Correlation with Target', 
             fontsize=14, fontweight='bold')
ax.axvline(x=0, color='black', linestyle='-', linewidth=0.8)
ax.grid(axis='x', alpha=0.3)
plt.tight_layout()
plt.show()

# ============================================================================
# 9.6 Compare Original vs Engineered Feature Importance
# ============================================================================
print("\n" + "=" * 80)
print("9.6 COMPARING ORIGINAL vs ENGINEERED FEATURES")
print("=" * 80)

# Get correlations for original features
original_feature_corr = X_train_eng_with_target[original_features + ['Diabetes_binary']].corr()['Diabetes_binary'].drop('Diabetes_binary')

# Compare top features
print("\nTop 10 Original Features (by absolute correlation):")
print("-" * 80)
top_original = original_feature_corr.abs().sort_values(ascending=False).head(10)
for feat, corr in top_original.items():
    print(f"  • {feat}: {original_feature_corr[feat]:.4f}")

print("\nTop 10 Engineered Features (by absolute correlation):")
print("-" * 80)
top_engineered = new_feature_corr.abs().sort_values(ascending=False).head(10)
for feat, corr in top_engineered.items():
    print(f"  • {feat}: {new_feature_corr[feat]:.4f}")

# Check if any engineered features outperform original features
max_original_corr = original_feature_corr.abs().max()
max_engineered_corr = new_feature_corr.abs().max()

print("\n" + "-" * 80)
print("COMPARISON SUMMARY:")
print("-" * 80)
print(f"\nMax correlation - Original features: {max_original_corr:.4f}")
print(f"Max correlation - Engineered features: {max_engineered_corr:.4f}")

if max_engineered_corr > max_original_corr:
    improvement = ((max_engineered_corr - max_original_corr) / max_original_corr) * 100
    print(f"\n✅ Engineered features show {improvement:.1f}% improvement in max correlation!")
else:
    print("\n📊 Original features still show strongest individual correlations")
    print("   However, engineered features may still improve model performance through:")
    print("   • Capturing non-linear interactions")
    print("   • Representing domain knowledge")
    print("   • Reducing feature engineering burden on the model")

# ============================================================================
# Summary & Variables Created
# ============================================================================
print("\n" + "=" * 80)
print("STEP 9 SUMMARY")
print("=" * 80)

print(f"\n✅ Feature Engineering Complete!")
print(f"\n📊 FEATURE COUNTS:")
print("-" * 80)
print(f"  • Original Features: {len(original_features)}")
print(f"  • Composite Features: {len(composite_features)}")
print(f"  • Interaction Features: {len(interaction_features)}")
print(f"  • Categorical Features: {len(categorical_features)}")
print(f"  • Total Engineered Features: {len(new_features)}")
print(f"  • TOTAL FEATURES: {len(X_train_engineered.columns)}")

print(f"\n📦 ENGINEERED DATASETS CREATED:")
print("-" * 80)
print(f"  • X_train_engineered: Original training set with engineered features")
print(f"  • X_train_smote_engineered: SMOTE training set with engineered features")
print(f"  • X_train_smotetomek_engineered: SMOTETomek with engineered features")
print(f"  • X_train_ros_engineered: Random Oversampling with engineered features")
print(f"  • X_test_engineered: Test set with engineered features")

print(f"\n🔬 FOR STEP 10 COMPARISON:")
print("-" * 80)
print("  We now have TWO versions of each dataset:")
print("  1. Original features only (21 features)")
print("  2. With engineered features (~{} features)".format(len(X_train_engineered.columns)))
print("\n  This allows us to compare:")
print("  • Models with vs without feature engineering")
print("  • Value added by domain knowledge")
print("  • Performance vs complexity trade-offs")

print("\n✅ Ready for Step 10: Baseline Model Training")
print("=" * 80)
```

```python
from sklearn.linear_model import LogisticRegression
from sklearn.tree import DecisionTreeClassifier
from sklearn.ensemble import RandomForestClassifier
from sklearn.metrics import (
    accuracy_score, precision_score, recall_score, f1_score, 
    roc_auc_score, confusion_matrix, classification_report,
    roc_curve, precision_recall_curve, average_precision_score
)
from sklearn.model_selection import cross_val_score
import time
import warnings
warnings.filterwarnings('ignore')

print("=" * 80)
print("STEP 10: BASELINE MODEL TRAINING & COMPARISON")
print("=" * 80)
print("\nTraining baseline models with comprehensive comparison:")
print("  • Multiple resampling techniques")
print("  • Original vs Engineered features")
print("  • Different model types")
print("\nThis will help identify the best approach for final modeling.\n")

# ============================================================================
# 10.1 Define Comparison Framework
# ============================================================================
print("=" * 80)
print("10.1 COMPARISON FRAMEWORK")
print("=" * 80)

print("\n📊 DATASETS TO COMPARE:")
print("-" * 80)

# Prepare all dataset combinations
datasets = {
    # Original Features (21 features)
    'Original + Class Weights': {
        'X_train': X_train_scaled,
        'y_train': y_train,
        'use_class_weight': True,
        'features': 'Original'
    },
    'SMOTE (Original Features)': {
        'X_train': X_train_smote,
        'y_train': y_train_smote,
        'use_class_weight': False,
        'features': 'Original'
    },
    'SMOTETomek (Original Features)': {
        'X_train': X_train_smotetomek,
        'y_train': y_train_smotetomek,
        'use_class_weight': False,
        'features': 'Original'
    },
    'Random Oversampling (Original)': {
        'X_train': X_train_ros,
        'y_train': y_train_ros,
        'use_class_weight': False,
        'features': 'Original'
    },
    
    # Engineered Features (39 features)
    'Engineered + Class Weights': {
        'X_train': X_train_engineered,
        'y_train': y_train,
        'use_class_weight': True,
        'features': 'Engineered'
    },
    'SMOTE (Engineered Features)': {
        'X_train': X_train_smote_engineered,
        'y_train': y_train_smote,
        'use_class_weight': False,
        'features': 'Engineered'
    },
    'SMOTETomek (Engineered Features)': {
        'X_train': X_train_smotetomek_engineered,
        'y_train': y_train_smotetomek,
        'use_class_weight': False,
        'features': 'Engineered'
    },
    'Random Oversampling (Engineered)': {
        'X_train': X_train_ros_engineered,
        'y_train': y_train_ros,
        'use_class_weight': False,
        'features': 'Engineered'
    }
}

for i, (name, info) in enumerate(datasets.items(), 1):
    X_shape = info['X_train'].shape if hasattr(info['X_train'], 'shape') else (len(info['X_train']), len(info['X_train'][0]))
    print(f"{i}. {name}")
    print(f"   • Training samples: {X_shape[0]:,}")
    print(f"   • Features: {X_shape[1]}")
    print(f"   • Class weights: {'Yes' if info['use_class_weight'] else 'No'}")

print("\n🤖 MODELS TO TEST:")
print("-" * 80)
print("1. Logistic Regression (Simple, Interpretable)")
print("2. Decision Tree (Non-linear, Interpretable)")
print("3. Random Forest (Ensemble, Robust)")

print("\n📈 EVALUATION METRICS:")
print("-" * 80)
print("• Accuracy: Overall correctness")
print("• Precision: Positive predictive value")
print("• Recall: Sensitivity (critical for diabetes detection)")
print("• F1-Score: Harmonic mean of precision and recall")
print("• ROC-AUC: Area under ROC curve")
print("• Training Time: Computational efficiency")

# ============================================================================
# 10.2 Define Evaluation Function
# ============================================================================
print("\n" + "=" * 80)
print("10.2 CREATING EVALUATION PIPELINE")
print("=" * 80)

def evaluate_model(model, X_train, y_train, X_test, y_test, model_name, dataset_name):
    """
    Train and evaluate a model, returning comprehensive metrics.
    """
    # Train model and measure time
    start_time = time.time()
    model.fit(X_train, y_train)
    train_time = time.time() - start_time
    
    # Make predictions
    y_pred = model.predict(X_test)
    y_pred_proba = model.predict_proba(X_test)[:, 1] if hasattr(model, 'predict_proba') else None
    
    # Calculate metrics
    metrics = {
        'Model': model_name,
        'Dataset': dataset_name,
        'Accuracy': accuracy_score(y_test, y_pred),
        'Precision': precision_score(y_test, y_pred, zero_division=0),
        'Recall': recall_score(y_test, y_pred, zero_division=0),
        'F1_Score': f1_score(y_test, y_pred, zero_division=0),
        'ROC_AUC': roc_auc_score(y_test, y_pred_proba) if y_pred_proba is not None else None,
        'Train_Time': train_time,
        'Train_Samples': len(X_train)
    }
    
    # Confusion matrix
    cm = confusion_matrix(y_test, y_pred)
    
    return metrics, model, y_pred, y_pred_proba, cm

print("\n✓ Evaluation pipeline ready!")

# ============================================================================
# 10.3 Train and Evaluate All Combinations
# ============================================================================
print("\n" + "=" * 80)
print("10.3 TRAINING BASELINE MODELS")
print("=" * 80)
print("\nThis may take a few minutes...")
print("-" * 80)

# Store all results
all_results = []
trained_models = {}

# Counter for progress
total_combinations = len(datasets) * 3  # 8 datasets × 3 models
current = 0

# Train all combinations
for dataset_name, dataset_info in datasets.items():
    X_train = dataset_info['X_train']
    y_train = dataset_info['y_train']
    use_class_weight = dataset_info['use_class_weight']
    feature_type = dataset_info['features']
    
    # Select appropriate test set
    X_test = X_test_engineered if feature_type == 'Engineered' else X_test_scaled
    
    print(f"\n📊 Dataset: {dataset_name}")
    print("-" * 80)
    
    # 1. Logistic Regression
    current += 1
    print(f"[{current}/{total_combinations}] Training Logistic Regression...")
    
    if use_class_weight:
        lr = LogisticRegression(max_iter=1000, random_state=42, class_weight='balanced', n_jobs=-1)
    else:
        lr = LogisticRegression(max_iter=1000, random_state=42, n_jobs=-1)
    
    lr_metrics, lr_model, lr_pred, lr_proba, lr_cm = evaluate_model(
        lr, X_train, y_train, X_test, y_test,
        'Logistic Regression', dataset_name
    )
    all_results.append(lr_metrics)
    trained_models[f"{dataset_name}_LR"] = {
        'model': lr_model, 'predictions': lr_pred, 
        'probabilities': lr_proba, 'confusion_matrix': lr_cm
    }
    print(f"   ✓ Accuracy: {lr_metrics['Accuracy']:.4f} | F1: {lr_metrics['F1_Score']:.4f} | ROC-AUC: {lr_metrics['ROC_AUC']:.4f}")
    
    # 2. Decision Tree
    current += 1
    print(f"[{current}/{total_combinations}] Training Decision Tree...")
    
    if use_class_weight:
        dt = DecisionTreeClassifier(max_depth=10, random_state=42, class_weight='balanced')
    else:
        dt = DecisionTreeClassifier(max_depth=10, random_state=42)
    
    dt_metrics, dt_model, dt_pred, dt_proba, dt_cm = evaluate_model(
        dt, X_train, y_train, X_test, y_test,
        'Decision Tree', dataset_name
    )
    all_results.append(dt_metrics)
    trained_models[f"{dataset_name}_DT"] = {
        'model': dt_model, 'predictions': dt_pred,
        'probabilities': dt_proba, 'confusion_matrix': dt_cm
    }
    print(f"   ✓ Accuracy: {dt_metrics['Accuracy']:.4f} | F1: {dt_metrics['F1_Score']:.4f} | ROC-AUC: {dt_metrics['ROC_AUC']:.4f}")
    
    # 3. Random Forest
    current += 1
    print(f"[{current}/{total_combinations}] Training Random Forest...")
    
    if use_class_weight:
        rf = RandomForestClassifier(n_estimators=100, max_depth=10, random_state=42, 
                                    class_weight='balanced', n_jobs=-1)
    else:
        rf = RandomForestClassifier(n_estimators=100, max_depth=10, random_state=42, n_jobs=-1)
    
    rf_metrics, rf_model, rf_pred, rf_proba, rf_cm = evaluate_model(
        rf, X_train, y_train, X_test, y_test,
        'Random Forest', dataset_name
    )
    all_results.append(rf_metrics)
    trained_models[f"{dataset_name}_RF"] = {
        'model': rf_model, 'predictions': rf_pred,
        'probabilities': rf_proba, 'confusion_matrix': rf_cm
    }
    print(f"   ✓ Accuracy: {rf_metrics['Accuracy']:.4f} | F1: {rf_metrics['F1_Score']:.4f} | ROC-AUC: {rf_metrics['ROC_AUC']:.4f}")

print("\n" + "=" * 80)
print("✅ All models trained successfully!")
print("=" * 80)

# ============================================================================
# 10.4 Results Summary
# ============================================================================
print("\n" + "=" * 80)
print("10.4 COMPREHENSIVE RESULTS COMPARISON")
print("=" * 80)

# Create results dataframe
results_df = pd.DataFrame(all_results)

# Sort by F1 Score (balanced metric)
results_df_sorted = results_df.sort_values('F1_Score', ascending=False)

print("\n📊 ALL RESULTS (Sorted by F1-Score):")
print("-" * 80)
display(results_df_sorted)

# ============================================================================
# 10.5 Best Models Analysis
# ============================================================================
print("\n" + "=" * 80)
print("10.5 TOP PERFORMING MODELS")
print("=" * 80)

# Top 5 by F1-Score
print("\n🏆 TOP 5 MODELS BY F1-SCORE:")
print("-" * 80)
top5_f1 = results_df_sorted.head(5)[['Model', 'Dataset', 'Accuracy', 'Precision', 'Recall', 'F1_Score', 'ROC_AUC']]
display(top5_f1)

# Top 5 by ROC-AUC
print("\n🏆 TOP 5 MODELS BY ROC-AUC:")
print("-" * 80)
top5_roc = results_df.sort_values('ROC_AUC', ascending=False).head(5)[['Model', 'Dataset', 'Accuracy', 'Precision', 'Recall', 'F1_Score', 'ROC_AUC']]
display(top5_roc)

# Top 5 by Recall (important for medical diagnosis)
print("\n🏆 TOP 5 MODELS BY RECALL (Diabetes Detection):")
print("-" * 80)
top5_recall = results_df.sort_values('Recall', ascending=False).head(5)[['Model', 'Dataset', 'Accuracy', 'Precision', 'Recall', 'F1_Score', 'ROC_AUC']]
display(top5_recall)

# ============================================================================
# 10.6 Comparison Visualizations
# ============================================================================
print("\n" + "=" * 80)
print("10.6 PERFORMANCE VISUALIZATIONS")
print("=" * 80)

# Create comparison plots
fig, axes = plt.subplots(2, 2, figsize=(18, 12))

# 1. F1-Score by Model Type
ax1 = axes[0, 0]
model_f1 = results_df.groupby('Model')['F1_Score'].mean().sort_values()
model_f1.plot(kind='barh', ax=ax1, color='steelblue', alpha=0.7)
ax1.set_xlabel('Average F1-Score', fontweight='bold')
ax1.set_title('Average F1-Score by Model Type', fontweight='bold', fontsize=12)
ax1.grid(axis='x', alpha=0.3)

# 2. ROC-AUC by Model Type
ax2 = axes[0, 1]
model_roc = results_df.groupby('Model')['ROC_AUC'].mean().sort_values()
model_roc.plot(kind='barh', ax=ax2, color='green', alpha=0.7)
ax2.set_xlabel('Average ROC-AUC', fontweight='bold')
ax2.set_title('Average ROC-AUC by Model Type', fontweight='bold', fontsize=12)
ax2.grid(axis='x', alpha=0.3)

# 3. Feature Engineering Impact
ax3 = axes[1, 0]
results_df['Feature_Type'] = results_df['Dataset'].apply(
    lambda x: 'Engineered' if 'Engineered' in x else 'Original'
)
feature_comparison = results_df.groupby('Feature_Type')[['Accuracy', 'F1_Score', 'ROC_AUC']].mean()
feature_comparison.plot(kind='bar', ax=ax3, alpha=0.7)
ax3.set_xlabel('Feature Type', fontweight='bold')
ax3.set_ylabel('Score', fontweight='bold')
ax3.set_title('Feature Engineering Impact', fontweight='bold', fontsize=12)
ax3.legend(['Accuracy', 'F1-Score', 'ROC-AUC'])
ax3.grid(axis='y', alpha=0.3)
ax3.set_xticklabels(ax3.get_xticklabels(), rotation=0)

# 4. Resampling Technique Impact
ax4 = axes[1, 1]
results_df['Resampling'] = results_df['Dataset'].apply(
    lambda x: 'SMOTE' if 'SMOTE' in x and 'Tomek' not in x
    else 'SMOTETomek' if 'SMOTETomek' in x
    else 'Random OS' if 'Random' in x
    else 'Class Weights'
)
resampling_comparison = results_df.groupby('Resampling')[['Accuracy', 'F1_Score', 'ROC_AUC']].mean()
resampling_comparison.plot(kind='bar', ax=ax4, alpha=0.7)
ax4.set_xlabel('Resampling Technique', fontweight='bold')
ax4.set_ylabel('Score', fontweight='bold')
ax4.set_title('Resampling Technique Impact', fontweight='bold', fontsize=12)
ax4.legend(['Accuracy', 'F1-Score', 'ROC-AUC'])
ax4.grid(axis='y', alpha=0.3)
ax4.set_xticklabels(ax4.get_xticklabels(), rotation=45, ha='right')

plt.tight_layout()
plt.show()

# ============================================================================
# 10.7 Detailed Analysis of Best Model
# ============================================================================
print("\n" + "=" * 80)
print("10.7 DETAILED ANALYSIS - BEST MODEL")
print("=" * 80)

# Get best model by F1-Score
best_model_row = results_df_sorted.iloc[0]
best_model_name = f"{best_model_row['Dataset']}_{best_model_row['Model'].replace(' ', '')}"

# Find the key in trained_models (handle spaces)
best_model_key = None
for key in trained_models.keys():
    if best_model_row['Dataset'] in key and best_model_row['Model'].replace(' ', '') in key:
        best_model_key = key
        break

if best_model_key:
    best_model_info = trained_models[best_model_key]
    
    print(f"\n🏆 BEST MODEL: {best_model_row['Model']}")
    print(f"📊 Dataset: {best_model_row['Dataset']}")
    print("-" * 80)
    print(f"Accuracy:  {best_model_row['Accuracy']:.4f}")
    print(f"Precision: {best_model_row['Precision']:.4f}")
    print(f"Recall:    {best_model_row['Recall']:.4f}")
    print(f"F1-Score:  {best_model_row['F1_Score']:.4f}")
    print(f"ROC-AUC:   {best_model_row['ROC_AUC']:.4f}")
    
    # Confusion Matrix
    print("\n📊 Confusion Matrix:")
    print("-" * 80)
    cm = best_model_info['confusion_matrix']
    cm_df = pd.DataFrame(cm, 
                         index=['Actual: No Diabetes', 'Actual: Diabetes'],
                         columns=['Predicted: No', 'Predicted: Yes'])
    display(cm_df)
    
    # Calculate additional metrics from confusion matrix
    tn, fp, fn, tp = cm.ravel()
    specificity = tn / (tn + fp)
    npv = tn / (tn + fn) if (tn + fn) > 0 else 0
    
    print("\n📈 Additional Metrics:")
    print("-" * 80)
    print(f"Specificity (True Negative Rate): {specificity:.4f}")
    print(f"Negative Predictive Value: {npv:.4f}")
    print(f"False Positive Rate: {fp / (fp + tn):.4f}")
    print(f"False Negative Rate: {fn / (fn + tp):.4f}")
    
    # Classification Report
    print("\n📋 Classification Report:")
    print("-" * 80)
    print(classification_report(y_test, best_model_info['predictions'], 
                               target_names=['No Diabetes', 'Diabetes']))
    
    # Visualize Confusion Matrix
    fig, axes = plt.subplots(1, 2, figsize=(14, 5))
    
    # Confusion Matrix Heatmap
    sns.heatmap(cm, annot=True, fmt='d', cmap='Blues', ax=axes[0],
                xticklabels=['No Diabetes', 'Diabetes'],
                yticklabels=['No Diabetes', 'Diabetes'])
    axes[0].set_title(f'Confusion Matrix - {best_model_row["Model"]}', 
                     fontweight='bold', fontsize=12)
    axes[0].set_ylabel('Actual', fontweight='bold')
    axes[0].set_xlabel('Predicted', fontweight='bold')
    
    # ROC Curve
    if best_model_info['probabilities'] is not None:
        fpr, tpr, _ = roc_curve(y_test, best_model_info['probabilities'])
        axes[1].plot(fpr, tpr, linewidth=2, label=f'ROC (AUC = {best_model_row["ROC_AUC"]:.4f})')
        axes[1].plot([0, 1], [0, 1], 'k--', linewidth=1, label='Random Classifier')
        axes[1].set_xlabel('False Positive Rate', fontweight='bold')
        axes[1].set_ylabel('True Positive Rate', fontweight='bold')
        axes[1].set_title('ROC Curve', fontweight='bold', fontsize=12)
        axes[1].legend()
        axes[1].grid(alpha=0.3)
    
    plt.tight_layout()
    plt.show()

# ============================================================================
# 10.8 Key Insights & Recommendations
# ============================================================================
print("\n" + "=" * 80)
print("10.8 KEY INSIGHTS & RECOMMENDATIONS")
print("=" * 80)

print("\n📊 PERFORMANCE SUMMARY:")
print("-" * 80)

# Feature Engineering Impact
orig_avg = results_df[results_df['Feature_Type'] == 'Original']['F1_Score'].mean()
eng_avg = results_df[results_df['Feature_Type'] == 'Engineered']['F1_Score'].mean()
fe_improvement = ((eng_avg - orig_avg) / orig_avg) * 100

print(f"\n1. Feature Engineering Impact:")
print(f"   • Original Features F1-Score: {orig_avg:.4f}")
print(f"   • Engineered Features F1-Score: {eng_avg:.4f}")
print(f"   • Improvement: {fe_improvement:+.2f}%")

if fe_improvement > 0:
    print(f"   ✅ Feature engineering IMPROVED performance")
else:
    print(f"   ⚠️  Feature engineering did not improve performance significantly")

# Model Type Performance
print(f"\n2. Model Type Performance:")
for model in results_df['Model'].unique():
    model_avg_f1 = results_df[results_df['Model'] == model]['F1_Score'].mean()
    model_avg_roc = results_df[results_df['Model'] == model]['ROC_AUC'].mean()
    print(f"   • {model}: F1={model_avg_f1:.4f}, ROC-AUC={model_avg_roc:.4f}")

# Resampling Technique Performance
print(f"\n3. Resampling Technique Performance:")
for technique in results_df['Resampling'].unique():
    tech_avg_f1 = results_df[results_df['Resampling'] == technique]['F1_Score'].mean()
    tech_avg_roc = results_df[results_df['Resampling'] == technique]['ROC_AUC'].mean()
    print(f"   • {technique}: F1={tech_avg_f1:.4f}, ROC-AUC={tech_avg_roc:.4f}")

print("\n" + "=" * 80)
print("✅ RECOMMENDATIONS FOR FINAL MODEL:")
print("=" * 80)
print(f"\n1. Best Overall Approach: {best_model_row['Model']} with {best_model_row['Dataset']}")
print(f"   • F1-Score: {best_model_row['F1_Score']:.4f}")
print(f"   • ROC-AUC: {best_model_row['ROC_AUC']:.4f}")
print(f"   • Recall: {best_model_row['Recall']:.4f} (important for diabetes detection)")

print("\n2. For Final Report:")
if fe_improvement > 2:
    print("   ✅ Include feature engineering - shows meaningful improvement")
else:
    print("   💡 Consider using original features for simplicity")

best_resampling = results_df.groupby('Resampling')['F1_Score'].mean().idxmax()
print(f"   ✅ Recommended resampling: {best_resampling}")

best_model_type = results_df.groupby('Model')['F1_Score'].mean().idxmax()
print(f"   ✅ Recommended model type: {best_model_type}")

print("\n3. Next Steps:")
print("   • Proceed to Step 11 for hyperparameter tuning on top performers")
print("   • Consider ensemble methods combining best models")
print("   • Perform cross-validation for robust performance estimates")

print("\n" + "=" * 80)
print("✅ Step 10: Baseline Model Training Complete!")
print("=" * 80)
```

```python
import shap
from sklearn.inspection import permutation_importance, PartialDependenceDisplay
from sklearn.model_selection import cross_val_score, cross_validate
import warnings
warnings.filterwarnings('ignore')

print("=" * 80)
print("STEP 12: MODEL INTERPRETATION & VALIDATION")
print("=" * 80)
print("\nAnalyzing TWO complementary models:")
print("  1. Random Forest + Class Weights (Primary) - Best F1-Score")
print("  2. Logistic Regression + Class Weights (Alternative) - Most Interpretable")
print("\n")

# ============================================================================
# 12.1 SECTION 1: PRIMARY MODEL - RANDOM FOREST INTERPRETATION
# ============================================================================
print("=" * 80)
print("12.1 PRIMARY MODEL: RANDOM FOREST + CLASS WEIGHTS")
print("=" * 80)
print(f"\nModel: {model2_results['Model']}")
print(f"Dataset: {model2_results['Dataset']}")
print(f"F1-Score: {model2_results['F1_Score']:.4f}")
print(f"ROC-AUC: {model2_results['ROC_AUC']:.4f}")
print("-" * 80)

# Get the best Random Forest model
rf_best_model = rf_random_2.best_estimator_

# ============================================================================
# 12.1.1 Feature Importance (Random Forest Built-in)
# ============================================================================
print("\n" + "-" * 80)
print("12.1.1 Feature Importance Analysis")
print("-" * 80)

# Get feature importances
feature_names = X_train_scaled.columns
feature_importances = rf_best_model.feature_importances_

# Create DataFrame
importance_df = pd.DataFrame({
    'Feature': feature_names,
    'Importance': feature_importances
}).sort_values('Importance', ascending=False)

print("\nTop 15 Most Important Features:")
display(importance_df.head(15))

# Visualize top features
fig, axes = plt.subplots(1, 2, figsize=(16, 6))

# Top 15 features
ax1 = axes[0]
top_15 = importance_df.head(15)
ax1.barh(range(len(top_15)), top_15['Importance'].values, color='steelblue', alpha=0.7)
ax1.set_yticks(range(len(top_15)))
ax1.set_yticklabels(top_15['Feature'].values)
ax1.set_xlabel('Importance', fontweight='bold')
ax1.set_title('Top 15 Features - Random Forest Importance', fontweight='bold', fontsize=12)
ax1.grid(axis='x', alpha=0.3)

# Cumulative importance
ax2 = axes[1]
cumulative_importance = importance_df['Importance'].cumsum()
ax2.plot(range(len(cumulative_importance)), cumulative_importance, linewidth=2)
ax2.axhline(y=0.95, color='r', linestyle='--', label='95% threshold')
ax2.set_xlabel('Number of Features', fontweight='bold')
ax2.set_ylabel('Cumulative Importance', fontweight='bold')
ax2.set_title('Cumulative Feature Importance', fontweight='bold', fontsize=12)
ax2.legend()
ax2.grid(alpha=0.3)

plt.tight_layout()
plt.show()

# Find number of features for 95% importance
n_features_95 = (cumulative_importance >= 0.95).argmax() + 1
print(f"\n💡 Top {n_features_95} features account for 95% of importance")

# ============================================================================
# 12.1.2 Permutation Importance
# ============================================================================
print("\n" + "-" * 80)
print("12.1.2 Permutation Importance (More Reliable)")
print("-" * 80)
print("\nCalculating permutation importance (this may take 1-2 minutes)...")

# Calculate permutation importance on test set
perm_importance = permutation_importance(
    rf_best_model, X_test_scaled, y_test,
    n_repeats=10, random_state=42, n_jobs=-1
)

# Create DataFrame
perm_importance_df = pd.DataFrame({
    'Feature': feature_names,
    'Importance_Mean': perm_importance.importances_mean,
    'Importance_Std': perm_importance.importances_std
}).sort_values('Importance_Mean', ascending=False)

print("\nTop 15 Features by Permutation Importance:")
display(perm_importance_df.head(15))

# Visualize
fig, ax = plt.subplots(figsize=(10, 8))
top_15_perm = perm_importance_df.head(15)
ax.barh(range(len(top_15_perm)), top_15_perm['Importance_Mean'].values, 
        xerr=top_15_perm['Importance_Std'].values,
        color='green', alpha=0.7, capsize=5)
ax.set_yticks(range(len(top_15_perm)))
ax.set_yticklabels(top_15_perm['Feature'].values)
ax.set_xlabel('Permutation Importance (with std)', fontweight='bold')
ax.set_title('Top 15 Features - Permutation Importance', fontweight='bold', fontsize=12)
ax.grid(axis='x', alpha=0.3)
plt.tight_layout()
plt.show()

# ============================================================================
# 12.1.3 SHAP Values (Global Interpretation)
# ============================================================================
print("\n" + "-" * 80)
print("12.1.3 SHAP Analysis (Global Feature Impact)")
print("-" * 80)
print("\nCalculating SHAP values (this may take 2-3 minutes)...")

# Use TreeExplainer for Random Forest
explainer = shap.TreeExplainer(rf_best_model)

# Calculate SHAP values on a sample (use 1000 samples for speed)
sample_size = min(1000, len(X_test_scaled))
X_test_sample = X_test_scaled.sample(n=sample_size, random_state=42)
shap_values = explainer.shap_values(X_test_sample)

# For binary classification, shap_values is a list [class_0, class_1]
# We want class_1 (diabetes)
shap_values_class1 = shap_values[1] if isinstance(shap_values, list) else shap_values

print("\n✓ SHAP values calculated successfully!")

# Summary plot
print("\nGenerating SHAP summary plot...")
plt.figure(figsize=(10, 8))
shap.summary_plot(shap_values_class1, X_test_sample, show=False)
plt.title('SHAP Summary Plot - Feature Impact on Diabetes Prediction', 
         fontweight='bold', fontsize=12, pad=20)
plt.tight_layout()
plt.show()

# Feature importance plot (mean absolute SHAP)
print("\nGenerating SHAP feature importance plot...")
plt.figure(figsize=(10, 8))
shap.summary_plot(shap_values_class1, X_test_sample, plot_type='bar', show=False)
plt.title('SHAP Feature Importance - Mean Absolute Impact', 
         fontweight='bold', fontsize=12, pad=20)
plt.tight_layout()
plt.show()

# ============================================================================
# 12.1.4 Partial Dependence Plots
# ============================================================================
print("\n" + "-" * 80)
print("12.1.4 Partial Dependence Plots (Feature Effects)")
print("-" * 80)

# Select top 6 features for PDP
top_6_features = importance_df.head(6)['Feature'].tolist()
print(f"\nCreating PDPs for top 6 features: {top_6_features}")

fig, axes = plt.subplots(2, 3, figsize=(18, 10))
axes = axes.flatten()

for idx, feature in enumerate(top_6_features):
    feature_idx = list(feature_names).index(feature)
    PartialDependenceDisplay.from_estimator(
        rf_best_model, X_test_scaled, [feature_idx],
        feature_names=feature_names,
        ax=axes[idx],
        grid_resolution=50
    )
    axes[idx].set_title(f'PDP: {feature}', fontweight='bold')

plt.suptitle('Partial Dependence Plots - Top 6 Features', 
            fontweight='bold', fontsize=14, y=1.00)
plt.tight_layout()
plt.show()

# ============================================================================
# 12.2 SECTION 2: ALTERNATIVE MODEL - LOGISTIC REGRESSION
# ============================================================================
print("\n" + "=" * 80)
print("12.2 ALTERNATIVE MODEL: LOGISTIC REGRESSION + CLASS WEIGHTS")
print("=" * 80)
print(f"\nModel: {model3_results['Model']}")
print(f"Dataset: {model3_results['Dataset']}")
print(f"F1-Score: {model3_results['F1_Score']:.4f}")
print(f"ROC-AUC: {model3_results['ROC_AUC']:.4f}")
print(f"Recall: {model3_results['Recall']:.4f} (Highest)")
print("-" * 80)

# Get the best Logistic Regression model
lr_best_model = lr_grid.best_estimator_

# ============================================================================
# 12.2.1 Coefficient Analysis
# ============================================================================
print("\n" + "-" * 80)
print("12.2.1 Logistic Regression Coefficients")
print("-" * 80)

# Get coefficients (engineered features)
feature_names_eng = X_train_engineered.columns
coefficients = lr_best_model.coef_[0]

# Create DataFrame
coef_df = pd.DataFrame({
    'Feature': feature_names_eng,
    'Coefficient': coefficients,
    'Abs_Coefficient': np.abs(coefficients)
}).sort_values('Abs_Coefficient', ascending=False)

print("\nTop 15 Features by Coefficient Magnitude:")
display(coef_df.head(15))

# Visualize
fig, axes = plt.subplots(1, 2, figsize=(16, 8))

# Positive and negative coefficients
ax1 = axes[0]
top_15_coef = coef_df.head(15)
colors = ['green' if x > 0 else 'red' for x in top_15_coef['Coefficient']]
ax1.barh(range(len(top_15_coef)), top_15_coef['Coefficient'].values, 
        color=colors, alpha=0.7)
ax1.set_yticks(range(len(top_15_coef)))
ax1.set_yticklabels(top_15_coef['Feature'].values)
ax1.set_xlabel('Coefficient Value', fontweight='bold')
ax1.set_title('Top 15 Features - Logistic Regression Coefficients', 
             fontweight='bold', fontsize=12)
ax1.axvline(x=0, color='black', linestyle='-', linewidth=0.8)
ax1.grid(axis='x', alpha=0.3)

# Odds ratios
ax2 = axes[1]
top_15_coef['Odds_Ratio'] = np.exp(top_15_coef['Coefficient'])
ax2.barh(range(len(top_15_coef)), top_15_coef['Odds_Ratio'].values, 
        color='steelblue', alpha=0.7)
ax2.set_yticks(range(len(top_15_coef)))
ax2.set_yticklabels(top_15_coef['Feature'].values)
ax2.set_xlabel('Odds Ratio', fontweight='bold')
ax2.set_title('Top 15 Features - Odds Ratios', fontweight='bold', fontsize=12)
ax2.axvline(x=1, color='black', linestyle='--', linewidth=1, label='No effect')
ax2.legend()
ax2.grid(axis='x', alpha=0.3)

plt.tight_layout()
plt.show()

# ============================================================================
# 12.2.2 Clinical Interpretation
# ============================================================================
print("\n" + "-" * 80)
print("12.2.2 Clinical Interpretation of Coefficients")
print("-" * 80)

print("\n📈 POSITIVE PREDICTORS (Increase Diabetes Risk):")
positive_coef = coef_df[coef_df['Coefficient'] > 0].head(10)
for idx, row in positive_coef.iterrows():
    odds_ratio = np.exp(row['Coefficient'])
    print(f"\n  • {row['Feature']}")
    print(f"    - Coefficient: +{row['Coefficient']:.4f}")
    print(f"    - Odds Ratio: {odds_ratio:.3f}")
    print(f"    - Impact: {(odds_ratio - 1) * 100:.1f}% increase in diabetes odds")

print("\n📉 NEGATIVE PREDICTORS (Decrease Diabetes Risk):")
negative_coef = coef_df[coef_df['Coefficient'] < 0].head(10)
for idx, row in negative_coef.iterrows():
    odds_ratio = np.exp(row['Coefficient'])
    print(f"\n  • {row['Feature']}")
    print(f"    - Coefficient: {row['Coefficient']:.4f}")
    print(f"    - Odds Ratio: {odds_ratio:.3f}")
    print(f"    - Impact: {(1 - odds_ratio) * 100:.1f}% decrease in diabetes odds")

# ============================================================================
# 12.3 SECTION 3: CROSS-VALIDATION (BOTH MODELS)
# ============================================================================
print("\n" + "=" * 80)
print("12.3 CROSS-VALIDATION FOR ROBUST EVALUATION")
print("=" * 80)

print("\nPerforming 5-fold cross-validation for both models...")
print("This may take 3-5 minutes...")
print("-" * 80)

# Define scoring metrics
scoring = {
    'accuracy': 'accuracy',
    'precision': 'precision',
    'recall': 'recall',
    'f1': 'f1',
    'roc_auc': 'roc_auc'
}

# CV for Random Forest
print("\n1. Random Forest + Class Weights (Original Features)...")
start_time = time.time()
rf_cv_scores = cross_validate(
    rf_best_model, X_train_scaled, y_train_clean,
    cv=5, scoring=scoring, n_jobs=-1, return_train_score=True
)
rf_cv_time = time.time() - start_time
print(f"   ✓ Completed in {rf_cv_time/60:.2f} minutes")

# CV for Logistic Regression
print("\n2. Logistic Regression + Class Weights (Engineered Features)...")
start_time = time.time()
lr_cv_scores = cross_validate(
    lr_best_model, X_train_engineered, y_train_clean,
    cv=5, scoring=scoring, n_jobs=-1, return_train_score=True
)
lr_cv_time = time.time() - start_time
print(f"   ✓ Completed in {lr_cv_time/60:.2f} minutes")

# Create results DataFrame
cv_results = pd.DataFrame({
    'Model': ['Random Forest', 'Logistic Regression'],
    'Accuracy_Mean': [
        rf_cv_scores['test_accuracy'].mean(),
        lr_cv_scores['test_accuracy'].mean()
    ],
    'Accuracy_Std': [
        rf_cv_scores['test_accuracy'].std(),
        lr_cv_scores['test_accuracy'].std()
    ],
    'Precision_Mean': [
        rf_cv_scores['test_precision'].mean(),
        lr_cv_scores['test_precision'].mean()
    ],
    'Precision_Std': [
        rf_cv_scores['test_precision'].std(),
        lr_cv_scores['test_precision'].std()
    ],
    'Recall_Mean': [
        rf_cv_scores['test_recall'].mean(),
        lr_cv_scores['test_recall'].mean()
    ],
    'Recall_Std': [
        rf_cv_scores['test_recall'].std(),
        lr_cv_scores['test_recall'].std()
    ],
    'F1_Mean': [
        rf_cv_scores['test_f1'].mean(),
        lr_cv_scores['test_f1'].mean()
    ],
    'F1_Std': [
        rf_cv_scores['test_f1'].std(),
        lr_cv_scores['test_f1'].std()
    ],
    'ROC_AUC_Mean': [
        rf_cv_scores['test_roc_auc'].mean(),
        lr_cv_scores['test_roc_auc'].mean()
    ],
    'ROC_AUC_Std': [
        rf_cv_scores['test_roc_auc'].std(),
        lr_cv_scores['test_roc_auc'].std()
    ]
})

print("\n📊 CROSS-VALIDATION RESULTS (5-Fold):")
print("-" * 80)
display(cv_results)

# Visualize CV results
fig, axes = plt.subplots(1, 2, figsize=(16, 6))

# Metrics comparison
ax1 = axes[0]
metrics = ['Accuracy', 'Precision', 'Recall', 'F1', 'ROC_AUC']
rf_means = [cv_results.loc[0, f'{m}_Mean'] for m in metrics]
lr_means = [cv_results.loc[1, f'{m}_Mean'] for m in metrics]

x = np.arange(len(metrics))
width = 0.35

ax1.bar(x - width/2, rf_means, width, label='Random Forest', alpha=0.7, color='steelblue')
ax1.bar(x + width/2, lr_means, width, label='Logistic Regression', alpha=0.7, color='green')

ax1.set_ylabel('Score', fontweight='bold')
ax1.set_title('Cross-Validation Scores Comparison', fontweight='bold', fontsize=12)
ax1.set_xticks(x)
ax1.set_xticklabels(metrics)
ax1.legend()
ax1.grid(axis='y', alpha=0.3)

# Standard deviation comparison (stability)
ax2 = axes[1]
rf_stds = [cv_results.loc[0, f'{m}_Std'] for m in metrics]
lr_stds = [cv_results.loc[1, f'{m}_Std'] for m in metrics]

ax2.bar(x - width/2, rf_stds, width, label='Random Forest', alpha=0.7, color='steelblue')
ax2.bar(x + width/2, lr_stds, width, label='Logistic Regression', alpha=0.7, color='green')

ax2.set_ylabel('Standard Deviation', fontweight='bold')
ax2.set_title('Model Stability (Lower is Better)', fontweight='bold', fontsize=12)
ax2.set_xticks(x)
ax2.set_xticklabels(metrics)
ax2.legend()
ax2.grid(axis='y', alpha=0.3)

plt.tight_layout()
plt.show()

# ============================================================================
# 12.4 SECTION 4: ERROR ANALYSIS
# ============================================================================
print("\n" + "=" * 80)
print("12.4 ERROR ANALYSIS - UNDERSTANDING MISCLASSIFICATIONS")
print("=" * 80)

# Analyze errors for both models
print("\n12.4.1 Random Forest Errors:")
print("-" * 80)

# Get predictions
rf_predictions = rf_best_model.predict(X_test_scaled)
rf_proba = rf_best_model.predict_proba(X_test_scaled)[:, 1]

# Identify errors
rf_errors = (rf_predictions != y_test)
rf_false_positives = ((rf_predictions == 1) & (y_test == 0))
rf_false_negatives = ((rf_predictions == 0) & (y_test == 1))

print(f"\nTotal Errors: {rf_errors.sum():,} ({rf_errors.sum()/len(y_test)*100:.2f}%)")
print(f"False Positives: {rf_false_positives.sum():,} (Predicted diabetes, actually healthy)")
print(f"False Negatives: {rf_false_negatives.sum():,} (Predicted healthy, actually diabetes)")

# Analyze error patterns
print("\n📊 Error Analysis - Feature Patterns:")

# Get indices of errors
fp_indices = np.where(rf_false_positives)[0]
fn_indices = np.where(rf_false_negatives)[0]

# Compare feature distributions
# Convert boolean Series to numpy array for proper indexing
correct_indices = np.where(~rf_errors)[0]

error_analysis = pd.DataFrame({
    'Feature': importance_df.head(10)['Feature'],
    'FP_Mean': [X_test_scaled.iloc[fp_indices][feat].mean() for feat in importance_df.head(10)['Feature']],
    'FN_Mean': [X_test_scaled.iloc[fn_indices][feat].mean() for feat in importance_df.head(10)['Feature']],
    'Correct_Mean': [X_test_scaled.iloc[correct_indices][feat].mean() for feat in importance_df.head(10)['Feature']]
})

display(error_analysis)

print("\n12.4.2 Logistic Regression Errors:")
print("-" * 80)

lr_predictions = lr_best_model.predict(X_test_engineered)
lr_proba = lr_best_model.predict_proba(X_test_engineered)[:, 1]

lr_errors = (lr_predictions != y_test)
lr_false_positives = ((lr_predictions == 1) & (y_test == 0))
lr_false_negatives = ((lr_predictions == 0) & (y_test == 1))

print(f"\nTotal Errors: {lr_errors.sum():,} ({lr_errors.sum()/len(y_test)*100:.2f}%)")
print(f"False Positives: {lr_false_positives.sum():,}")
print(f"False Negatives: {lr_false_negatives.sum():,}")

# ============================================================================
# 12.5 FINAL SUMMARY & RECOMMENDATIONS
# ============================================================================
print("\n" + "=" * 80)
print("12.5 FINAL SUMMARY & RECOMMENDATIONS")
print("=" * 80)

print("\n🏆 PRIMARY MODEL: Random Forest + Class Weights")
print("-" * 80)
print(f"✓ Test F1-Score: {model2_results['F1_Score']:.4f}")
print(f"✓ Test ROC-AUC: {model2_results['ROC_AUC']:.4f}")
print(f"✓ CV F1-Score: {cv_results.loc[0, 'F1_Mean']:.4f} ± {cv_results.loc[0, 'F1_Std']:.4f}")
print(f"✓ Top Features: {', '.join(importance_df.head(3)['Feature'].tolist())}")
print("\n🎯 Use Case: Production deployment - best balanced performance")

print("\n📊 ALTERNATIVE MODEL: Logistic Regression + Class Weights")
print("-" * 80)
print(f"✓ Test F1-Score: {model3_results['F1_Score']:.4f}")
print(f"✓ Test ROC-AUC: {model3_results['ROC_AUC']:.4f}")
print(f"✓ Test Recall: {model3_results['Recall']:.4f} (HIGHEST)")
print(f"✓ CV F1-Score: {cv_results.loc[1, 'F1_Mean']:.4f} ± {cv_results.loc[1, 'F1_Std']:.4f}")
print("\n🎯 Use Case: Screening programs - maximizes diabetes detection")

print("\n💡 KEY INSIGHTS:")
print("-" * 80)
print("1. Both models show stable performance across CV folds")
print("2. Random Forest provides best F1-Score balance")
print("3. Logistic Regression captures most diabetes cases (high recall)")
print("4. Top features consistently important across models:")
for i, feat in enumerate(importance_df.head(5)['Feature'], 1):
    print(f"   {i}. {feat}")

print("\n✅ FOR FINAL REPORT:")
print("-" * 80)
print("• Present Random Forest as PRIMARY model")
print("• Include Logistic Regression for interpretability")
print("• Highlight SHAP analysis for explainability")
print("• Discuss coefficient interpretation for clinical relevance")
print("• Show cross-validation results for robustness")

print("\n" + "=" * 80)
print("✅ Step 12: Model Interpretation & Validation Complete!")
print("=" * 80)
```
